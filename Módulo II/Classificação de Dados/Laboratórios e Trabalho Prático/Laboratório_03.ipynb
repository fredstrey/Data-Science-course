```python
# Escreva sua resposta aqui
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier

class ClassifierPipeline:
    def __init__(self):
        self.classifiers = {
            'KNN': KNeighborsClassifier(),
            'DecisionTree': DecisionTreeClassifier(),
            'NaiveBayes': GaussianNB(),
            'LogisticRegression': LogisticRegression(),
            'AdaBoost': AdaBoostClassifier(),
            'RandomForest': RandomForestClassifier()
        }

    def fit_classifiers(self, X_train, y_train):
        self.trained_classifiers = {}
        for name, clf in self.classifiers.items():
            clf.fit(X_train, y_train)
            self.trained_classifiers[name] = clf

    def predict_classifiers(self, X_test):
        predictions = {}
        for name, clf in self.trained_classifiers.items():
            predictions[name] = clf.predict(X_test)
        return predictions

pipeline = ClassifierPipeline()
pipeline.fit_classifiers(X_train, y_train)

# Realizando previsões
predictions = pipeline.predict_classifiers(X_test)
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.metrics import roc_curve, auc, classification_report
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from joblib import Parallel, delayed
from sklearn.model_selection import StratifiedKFold


class ClassifierOptimizer:
    def __init__(self, X_train, y_train, X_test, y_test):
        if X_train.shape[0] != y_train.shape[0]:
            raise ValueError("X_train e y_train devem ter o mesmo número de amostras.")
        if X_test.shape[0] != y_test.shape[0]:
            raise ValueError("X_test e y_test devem ter o mesmo número de amostras.")
        self.X_train = X_train
        self.y_train = y_train
        self.X_test = X_test
        self.y_test = y_test
        self.scaler = None
        self.scaled = False
        self.cv=5
        self.classifiers = {
            'KNN': KNeighborsClassifier(),
            'LogisticRegression': LogisticRegression(),
            'DecisionTree': DecisionTreeClassifier(),
            'NaiveBayes': GaussianNB(),
            'AdaBoost': AdaBoostClassifier(),
            'RandomForest': RandomForestClassifier()
        }
        self.param_grids = {
            'KNN': {'n_neighbors': np.arange(1, 31)},
            'LogisticRegression': {'C': np.logspace(-3, 3, 30)},
            'DecisionTree': {'max_depth': np.arange(2, 31, 2)},
            'NaiveBayes': {'var_smoothing': np.logspace(-11, 11, 30)},
            'AdaBoost': {'n_estimators': [10, 100, 500, 1000]},
            'RandomForest': {
                'n_estimators': [10, 100, 500, 1000],
                'max_depth': np.arange(2, 31, 2)
            }
        }
        self.best_estimators = {}
        self.best_hyperparameters = {}
        self.predictions_optimized = {}
        self.class_predictions = {}

    def scale_data(self):
        """
        Aplica o MinMaxScaler aos dados de treino e teste.
        """
        self.scaler = MinMaxScaler()
        self.X_train_scaled = self.scaler.fit_transform(self.X_train)
        self.X_test_scaled = self.scaler.transform(self.X_test)
        self.scaled = True
        print("Dados escalados com sucesso!")

    def get_training_data(self):
        """
        Retorna os dados de treino escalados (se aplicável) ou não escalados.
        """
        return self.X_train_scaled if self.scaled else self.X_train

    def get_testing_data(self):
        """
        Retorna os dados de teste escalados (se aplicável) ou não escalados.
        """
        return self.X_test_scaled if self.scaled else self.X_test

    def optimize_classifiers(self):
        """
        Otimiza os classificadores usando GridSearchCV.
        """
        X_train = self.get_training_data()
        for classifier_name, classifier in self.classifiers.items():
            print(f"\nOtimizando {classifier_name}...")
            
            grid_search = GridSearchCV(
                classifier, 
                self.param_grids[classifier_name], 
                cv=StratifiedKFold(n_splits=self.cv, shuffle=True, random_state=42), 
                n_jobs=-1, 
                verbose=1
            )
            grid_search.fit(X_train, self.y_train)
            
            self.best_estimators[classifier_name] = grid_search.best_estimator_
            self.best_hyperparameters[classifier_name] = grid_search.best_params_
            print(f"Melhores parâmetros para {classifier_name}: {grid_search.best_params_}")

    def evaluate_classifier(self, classifier_name, classifier, X_train, X_test, y_train):
        """
        Avalia um único classificador.
        """
        print(f"Avaliando {classifier_name}...")
        classifier.fit(X_train, y_train)
        
        # Previsões probabilísticas
        if hasattr(classifier, "predict_proba"):
            y_pred_proba = classifier.predict_proba(X_test)[:, 1]
        else:
            y_pred_proba = None
        
        # Previsões de classe
        y_pred = classifier.predict(X_test)
        
        return classifier_name, y_pred_proba, y_pred

    def evaluate_classifiers(self):
        """
        Avalia os classificadores nos dados de teste usando paralelização com joblib.
        """
        X_train = self.get_training_data()
        X_test = self.get_testing_data()
        
        if not self.best_estimators:
            print("Usando classificadores padrão (sem otimização).")
            self.best_estimators = self.classifiers
        
        # Usar joblib.Parallel para paralelizar
        results = Parallel(n_jobs=-1)(
            delayed(self.evaluate_classifier)(
                classifier_name,
                classifier,
                X_train,
                X_test,
                self.y_train,
            )
            for classifier_name, classifier in self.best_estimators.items()
        )
        
        # Coletar resultados
        for classifier_name, y_pred_proba, y_pred in results:
            self.predictions_optimized[classifier_name] = y_pred_proba
            self.class_predictions[classifier_name] = y_pred

    def print_best_hyperparameters(self):
        """
        Imprime os melhores hiperparâmetros encontrados para cada classificador.
        """
        if not self.best_hyperparameters:
            raise ValueError("Chame optimize_classifiers() antes de imprimir os hiperparâmetros.")
        
        print("\nMelhores Hiperparâmetros:")
        for classifier_name, params in self.best_hyperparameters.items():
            print(f"{classifier_name}: {params}")

    def print_classification_reports(self):
        """
        Imprime relatórios de classificação para todos os classificadores.
        """
        if not self.class_predictions:
            raise ValueError("Chame evaluate_classifiers() antes de gerar relatórios.")
        
        for classifier_name, y_pred in self.class_predictions.items():
            print(f"\n{classifier_name}\n{'-'*30}")
            print(classification_report(self.y_test, y_pred))

    def print_auc_scores(self):
        """
        Imprime as métricas AUC para todos os classificadores.
        """
        if not self.predictions_optimized:
            raise ValueError("Chame evaluate_classifiers() antes de calcular AUC.")
        
        for classifier_name, y_pred_proba in self.predictions_optimized.items():
            fpr, tpr, _ = roc_curve(self.y_test, y_pred_proba)
            roc_auc = auc(fpr, tpr)
            print(f"AUC ({classifier_name}): {roc_auc:.4f}")

    def plot_roc_curve(self):
        """
        Plota todas as curvas ROC em um único gráfico.
        """
        if not self.predictions_optimized:
            raise ValueError("Chame evaluate_classifiers() antes de plotar curvas ROC.")
        
        plt.figure(figsize=(10, 8))
        for classifier_name, y_pred_proba in self.predictions_optimized.items():
            fpr, tpr, _ = roc_curve(self.y_test, y_pred_proba)
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f'{classifier_name} (AUC = {roc_auc:.2f})')
        
        plt.plot([0, 1], [0, 1], 'k--', label='Aleatório')
        plt.title('Curvas ROC Comparativas')
        plt.xlabel('Taxa de Falsos Positivos')
        plt.ylabel('Taxa de Verdadeiros Positivos')
        plt.legend(loc='lower right')
        plt.show()

optimizer = ClassifierOptimizer(X_train, y_train, X_test, y_test)
optimizer.evaluate_classifiers()
```

```python
# Escreva sua resposta aqui
from sklearn.metrics import roc_curve, auc

def compute_auc(predictions, y_test):
    # Iterar pelos classificadores e calcular a AUC
    auc_scores = {}
    for classifier, y_pred in predictions.items():
        # Calcular FPR e TPR para a curva ROC
        fpr, tpr, _ = roc_curve(y_test, y_pred)
        
        # Calcular AUC
        roc_auc = auc(fpr, tpr)
        
        # Armazenar o valor da AUC no dicionário
        auc_scores[classifier] = roc_auc
    
    # Imprimir a AUC para cada classificador
    for classifier, auc_score in auc_scores.items():
        print(f'{classifier}: AUC = {auc_score:.2f}')

# Exemplo de uso:
compute_auc(predictions, y_test)
```

### Item 2.1

Para isso, você deve utilizar o módulo [`GridSearchCV`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html).

Lembre, nenhuma informação do conjunto de teste deve ser utilizada na fase de treinamento e validação cruzada.

Você deve considerar os seis classificadores vistos até o momento (e já utilizados neste Laboratório). Para cada um deles, considere os seguintes conjuntos de hiperparâmetros candidatos:
- $k$ do **KNN** -- $k \in \{1, 2, 3, \dots, 30\}$.
- $C$ da **Regressão Logística** -- comumente, para isso, utiliza-se valores espaçados exponencialmente. Assim, utilize a função [`logspace`](https://numpy.org/doc/stable/reference/generated/numpy.logspace.html) da biblioteca `numpy` para criar 30 valores espaçados exponencialmente de 0.001 a 1000.
- Profundidade da **Árvore de Decisão** -- considere os valores em $\{2, 4, 6, \dots, 30\}$.
- `var_smoothin` do **Gaussian Naive Bayes** -- comumente, para isso, utiliza-se valores espaçados exponencialmente. Assim, utilize a função [`logspace`](https://numpy.org/doc/stable/reference/generated/numpy.logspace.html) da biblioteca `numpy` para criar 30 valores espaçados exponencialmente de $10^{-11}$ a $10^{11}$.
- número de estimadores do **AdaBoost** -- considere os valores $\{10, 100, 500, 1000\}$.
- número de estimadores e profundidade máxima da **Floresta Aleatória** -- considere os valores $\{10, 100, 500, 1000\}$ para o número de estimadores e $\{2, 4, 6, \dots, 30\}$ para profundidade máxima.

Nesse momento, vocês devem realizar a validação cruzada e treinamento dos modelos.

**Observação:** Na Parte 2 do Laboratório, vocês devem usar os dados transformados por reescala (i.e., `X_train_scaled` e `X_test_scaled`).

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
import numpy as np
from sklearn.metrics import classification_report, roc_curve, auc

class ClassifierOptimizer:
    def __init__(self, X_train_scaled, y_train, X_test_scaled, y_test):
        self.X_train_scaled = X_train_scaled
        self.y_train = y_train
        self.X_test_scaled = X_test_scaled
        self.y_test = y_test
        self.classifiers = {
            'KNN': KNeighborsClassifier(),
            'LogisticRegression': LogisticRegression(),
            'DecisionTree': DecisionTreeClassifier(),
            'NaiveBayes': GaussianNB(),
            'AdaBoost': AdaBoostClassifier(),
            'RandomForest': RandomForestClassifier()
        }
        self.param_grids = {
            'KNN': {'n_neighbors': np.arange(1, 31)},
            'LogisticRegression': {'C': np.logspace(-3, 3, 30)},
            'DecisionTree': {'max_depth': np.arange(2, 31, 2)},
            'NaiveBayes': {'var_smoothing': np.logspace(-11, 11, 30)},
            'AdaBoost': {'n_estimators': [10, 100, 500, 1000]},
            'RandomForest': {
                'n_estimators': [10, 100, 500, 1000],
                'max_depth': np.arange(2, 31, 2)
            }
        }
        self.best_estimators = {}
        self.predictions_optimized = {}

    def optimize_classifiers(self):
        for classifier_name, classifier in self.classifiers.items():
            print(f"Otimizando {classifier_name}...")

            # Realizar GridSearchCV
            grid_search = GridSearchCV(classifier, self.param_grids[classifier_name], cv=5, n_jobs=-1, verbose=1)
            grid_search.fit(self.X_train_scaled, self.y_train)

            # Armazenar o melhor estimador
            self.best_estimators[classifier_name] = grid_search.best_estimator_
            print(f"Melhores parâmetros para {classifier_name}: {grid_search.best_params_}")
            print(f"Melhor pontuação de validação cruzada: {grid_search.best_score_:.4f}\n")

    def evaluate_classifiers(self):
        for classifier_name, best_estimator in self.best_estimators.items():
            # Fazer previsões com o melhor estimador
            self.predictions_optimized[classifier_name] = best_estimator.predict(self.X_test_scaled)
        
        # Exibir relatórios de classificação para os melhores modelos
        for classifier_name, y_pred in self.predictions_optimized.items():
            print(f'\t\t{classifier_name}\n', classification_report(self.y_test, y_pred))

    def compute_auc(self):
        auc_scores = {}
        for classifier_name, y_pred in self.predictions_optimized.items():
            fpr, tpr, _ = roc_curve(self.y_test, y_pred)
            roc_auc = auc(fpr, tpr)
            auc_scores[classifier_name] = roc_auc
        
        for classifier_name, auc_score in auc_scores.items():
            print(f'{classifier_name}: AUC = {auc_score:.2f}')

# Exemplo de como utilizar a classe:
# Suponha que você tenha X_train_scaled, X_test_scaled, y_train, y_test já definidos

optimizer = ClassifierOptimizer(X_train_scaled, y_train, X_test_scaled, y_test)
optimizer.optimize_classifiers()
optimizer.evaluate_classifiers()
optimizer.compute_auc()
```

### Item 2.5

Compute e imprima a [Área sob a Curva Roc](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.auc.html) para os seis classificadores.

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

class ClassifierOptimizer:
    def __init__(self, X_train_scaled, y_train, X_test_scaled, y_test):
        self.X_train_scaled = X_train_scaled
        self.y_train = y_train
        self.X_test_scaled = X_test_scaled
        self.y_test = y_test
        self.classifiers = {
            'KNN': KNeighborsClassifier(),
            'LogisticRegression': LogisticRegression(),
            'DecisionTree': DecisionTreeClassifier(),
            'NaiveBayes': GaussianNB(),
            'AdaBoost': AdaBoostClassifier(),
            'RandomForest': RandomForestClassifier()
        }
        self.param_grids = {
            'KNN': {'n_neighbors': np.arange(1, 31)},
            'LogisticRegression': {'C': np.logspace(-3, 3, 30)},
            'DecisionTree': {'max_depth': np.arange(2, 31, 2)},
            'NaiveBayes': {'var_smoothing': np.logspace(-11, 11, 30)},
            'AdaBoost': {'n_estimators': [10, 100, 500, 1000]},
            'RandomForest': {
                'n_estimators': [10, 100, 500, 1000],
                'max_depth': np.arange(2, 31, 2)
            }
        }
        self.best_estimators = {}
        self.predictions_optimized = {}

    def optimize_classifiers(self):
        """
        Realiza a otimização dos classificadores usando GridSearchCV.
        Ajusta os hiperparâmetros para os modelos e seleciona o melhor estimador.
        """
        for classifier_name, classifier in self.classifiers.items():
            print(f"Otimizando {classifier_name}...")

            # Realizar GridSearchCV
            grid_search = GridSearchCV(classifier, self.param_grids[classifier_name], cv=5, n_jobs=-1, verbose=1)
            grid_search.fit(self.X_train_scaled, self.y_train)

            # Armazenar o melhor estimador encontrado
            self.best_estimators[classifier_name] = grid_search.best_estimator_

    def evaluate_classifiers(self):
        """
        Avalia os classificadores otimizados no conjunto de teste e armazena as previsões.
        """
        for classifier_name, best_estimator in self.best_estimators.items():
            # Fazer previsões com o melhor estimador
            y_pred_proba = best_estimator.predict_proba(self.X_test_scaled)[:, 1]
            self.predictions_optimized[classifier_name] = y_pred_proba

    def print_auc_scores(self):
        """
        Calcula e imprime a AUC para cada classificador.
        """
        for classifier_name, y_pred_proba in self.predictions_optimized.items():
            # Calcular a curva ROC
            fpr, tpr, _ = roc_curve(self.y_test, y_pred_proba)
            # Calcular a AUC (Área sob a Curva)
            roc_auc = auc(fpr, tpr)
            print(f"AUC para {classifier_name}: {roc_auc:.4f}")

    def plot_roc_curve(self):
        """
        Plota a curva ROC para todos os classificadores.
        """
        plt.figure(figsize=(10, 8))
        for classifier_name, y_pred_proba in self.predictions_optimized.items():
            # Calcular a curva ROC
            fpr, tpr, _ = roc_curve(self.y_test, y_pred_proba)
            # Calcular a AUC (Área sob a Curva)
            roc_auc = auc(fpr, tpr)
            # Plotar a curva ROC
            plt.plot(fpr, tpr, label=f'{classifier_name} (AUC = {roc_auc:.2f})')

        # Adicionar linha diagonal para referência
        plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')

        # Definir título e labels
        plt.title('Curva ROC para Classificadores')
        plt.xlabel('Taxa de Falsos Positivos (FPR)')
        plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
        plt.legend(loc='lower right')

        plt.show()

# Exemplo de como usar a classe:

# Suponha que você tenha os dados pré-processados:
# X_train_scaled, X_test_scaled, y_train, y_test

optimizer = ClassifierOptimizer(X_train_scaled, y_train, X_test_scaled, y_test)
optimizer.optimize_classifiers()  # Realiza a otimização dos classificadores
optimizer.evaluate_classifiers()  # Avalia os classificadores no conjunto de teste
optimizer.print_auc_scores()  # Imprime as AUCs
optimizer.plot_roc_curve()  # Plota a curva ROC para todos os classificadores
```
