```python
numericas = [
   'Year_Birth', 'Income', 'Kidhome', 'Teenhome',
    'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
    'NumStorePurchases', 'NumWebVisitsMonth', 'AcceptedCmp3',
    'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2',
    'Complain', 'Z_CostContact', 'Z_Revenue', 'Response'
]

categoricas = ['Education', 'Marital_Status']
```

```python
df = df.drop(columns=['Dt_Customer'])
```

```python
print(df['Education'].unique())

```

```python
df_dbscan= df.copy()
```

```python
education_order = {
    'Basic': 0,
    '2n Cycle': 1,
    'Graduation': 2,
    'Master': 3,
    'PhD': 4
}

df_dbscan['Education_encoded'] = df_dbscan['Education'].map(education_order)
```

```python
print(df_dbscan['Marital_Status'].unique())
```

```python
df_dbscan = pd.get_dummies(df, columns=['Marital_Status'], prefix='Marital', drop_first=False)
```

```python
df_dbscan[numericas].isna().sum()
```

```python
df = df.dropna(subset=['Income'])
df.reset_index(drop=True, inplace=True)
df_dbscan = df_dbscan.dropna(subset=['Income'])
df_dbscan.reset_index(drop=True, inplace=True)
df_dbscan = df_dbscan.drop(['Education'], axis=1).copy()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_dbscan[numericas] = scaler.fit_transform(df_dbscan[numericas])
dbscan = DBSCAN(eps=1.5, min_samples=5) 
df_dbscan['cluster_dbscan'] = dbscan.fit_predict(df_dbscan[numericas])
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np


melhor_score = -1
melhor_eps = None
melhor_min_samples = None


for eps in np.arange(0.5, 5.1, 0.1):  # eps de 1.0 a 5.0
    for min_samples in range(3, 10):  # min_samples de 3 a 9
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(df_dbscan[numericas])

        # Verificar se tem pelo menos 2 clusters (excluindo ru√≠do -1)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        if n_clusters >= 2:
            score = silhouette_score(df_dbscan[numericas], labels)
            print(f"eps={eps}, min_samples={min_samples} ‚Üí clusters={n_clusters}, silhouette={score:.4f}")

            if score > melhor_score:
                melhor_score = score
                melhor_eps = eps
                melhor_min_samples = min_samples

print("\nüîç Melhor configura√ß√£o:")
print(f"eps={melhor_eps}, min_samples={melhor_min_samples}, silhouette={melhor_score:.4f}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

colunas_numericas = df_dbscan.select_dtypes(include='float64').columns
colunas_numericas = colunas_numericas.drop('ID', errors='ignore')

# Filtra colunas com vari√¢ncia maior que 0.01
variancia = df_dbscan[colunas_numericas].var()
colunas_boas = variancia[variancia > 0.01].index

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_dbscan[colunas_boas])

# Aplica PCA com muitos componentes (at√© o m√°ximo poss√≠vel)
n_componentes = X_scaled.shape[1]
pca = PCA(n_components=n_componentes)
X_pca = pca.fit_transform(X_scaled)

# Vari√¢ncia explicada acumulada
var_acumulada = np.cumsum(pca.explained_variance_ratio_)

# Gr√°fico
plt.figure(figsize=(10, 6))
plt.plot(var_acumulada, marker='o')
plt.title("Vari√¢ncia Explicada Acumulada pelo PCA")
plt.xlabel("N√∫mero de Componentes Principais")
plt.ylabel("Vari√¢ncia Explicada Acumulada")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import RobustScaler

df = pd.read_csv('marketing_campaign.csv', sep='\t')
df = df.drop(columns=['ID', 'Z_CostContact', 'Z_Revenue', 'Dt_Customer']).dropna(subset=['Income'])

#vamos juntar algumas colunas para facilitar a an√°lise

df['Age'] = 2023 - df['Year_Birth']  # Transformar ano em idade
df['Family_Size'] = df['Kidhome'] + df['Teenhome'] + 1  # unificar todas as crian√ßas em uma s√≥ coluna
df['Total_Spending'] = df.filter(like='Mnt').sum(axis=1)  # junta os gastos por categoria em uma s√≥ coluna

# Codificar categ√≥ricas
df['Education'] = df['Education'].map({'Basic':0, '2n Cycle':1, 'Graduation':2, 'Master':3, 'PhD':4})
df = pd.get_dummies(df, columns=['Marital_Status'], drop_first=True)

# Normaliza√ß√£o robusta (para lidar com outliers), RobustScaler usa mediana e Intervalo Interquartil (IQR) para normalizar os dados, sendo menos sens√≠vel a outliers
scaler = RobustScaler()
X = scaler.fit_transform(df.select_dtypes(include=['number']))
```

```python
from sklearn.manifold import trustworthiness
import umap
trust_scores = []
n_components_range = range(2, 15)

for n in n_components_range:
# Vamos avaliar se os 15 vizinhos (n_neighbors=15) mais pr√≥ximos de cada ponto no espa√ßo original continuam sendo vizinhos no espa√ßo projetado.
    reducer = umap.UMAP(n_components=n, random_state=42)
    X_embedding = reducer.fit_transform(X)
    trust_scores.append(trustworthiness(X, X_embedding, n_neighbors=15)) # trustworthiness usa knn para comparar os vizinhos mais pr√≥ximos no espa√ßo original e no espa√ßo projetado.

plt.plot(n_components_range, trust_scores, 'go-')
plt.xlabel('N√∫mero de Componentes UMAP')
plt.ylabel('Confian√ßa Geom√©trica (0-1)')
plt.title('Preserva√ß√£o da Estrutura vs. N¬∫ de Componentes')
plt.grid(True)
```

5 componentes projetadas j√° possuem boa confian√ßa (~0.975), 97,5% dos vizinhos mais pr√≥ximos de cada ponto no espa√ßo original tamb√©m aparecem entre os vizinhos mais pr√≥ximos no espa√ßo projetado

```python
import umap
# Aplicando UMAP com 5 componentes, pois j√° preserva a estrutura
reducer = umap.UMAP(n_components=5, random_state=42, n_neighbors=15)
X_embedding = reducer.fit_transform(X)
```

```python
from sklearn.cluster import DBSCAN

# DBSCAN para identificar ru√≠do e outliers
dbscan = DBSCAN(eps=3.5, min_samples=10)
noise_mask = dbscan.fit_predict(X_embedding) == -1  # Identifica pontos de ru√≠do (dbsacn classifica outliers como -1)

# Remover outliers
X_clean = X_embedding[~noise_mask]
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# M√©todo do Cotovelo
inertia = []
k_range = range(2, 10)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_clean)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, 'bo-', markersize=8, linewidth=2)
plt.title('M√©todo do Cotovelo para Determina√ß√£o de K Ideal', fontsize=14)
plt.xlabel('N√∫mero de Clusters (K)', fontsize=12)
plt.ylabel('In√©rcia (Within-Cluster Sum of Squares)', fontsize=12)
plt.xticks(k_range)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# K=4
kmeans_final = KMeans(n_clusters=4, random_state=42)
clusters = kmeans_final.fit_predict(X_clean)

print(f"Silhouette Score: {silhouette_score(X_clean, clusters):.3f}")
print(f"Davies-Bouldin Score: {davies_bouldin_score(X_clean, clusters):.3f}")
```

```python
df_clean = df[~noise_mask].copy() #remove ouliers
df_clean['Cluster'] = clusters
#valores m√©dios por cluster
cluster_profile = df_clean.groupby('Cluster').mean()
print(cluster_profile[['Age', 'Income', 'Total_Spending', 'NumWebPurchases']])
```
