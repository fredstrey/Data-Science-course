## Exerccio 6.1.

Dados os pontos cartesianos $(X_1, Y_1)$ e $(X_2, Y_2)$ como entrada pelo teclado, escreva um cdigo que exiba na tela a distncia entre ambos.

**Observao:** A distncia $AB$ entre dois pontos $A = (X_a, Y_a)$ e $B = (X_b, Y_b)$  dada por $|AB| = \sqrt{{(X_b - X_a)}^{2} + {(Y_b - Y_a)}^{2}}$.

**Entrada:**\
0\
0\
1\
1\
**Sada:**\
1.4142135623730951

````python
import math
def calcular_distancia():
  x1 = float(input('digite x1'))
  y1 = float(input('digite y1'))
  x2 = float(input('digite x2'))
  y2 = float(input('digite y2'))
  distancia = math.sqrt((x2-x1)**2+(y2-y1)**2)
  print(distancia)
calcular_distancia()
````

## Exerccio 6.2.

Escreva um cdigo que receba duas string, a primeira sendo uma srie de caracteres e a segunda a palavra que voc deseja encontrar naquela sequncia (o cdigo deve desconsiderar diferenas entre letras maisculas e minsculas e deve imprimir a palavra encontrada em CAIXA ALTA):

**Entrada:**\
abksjdIntrocompaksdj\
introcomp\
**Sada:**\
Palavra encontrada: INTROCOMP\
\
**Entrada:**\
abksabksabks\
introcomp\
**Sada:**\
Palavra no encontrada!

````python
def decifrar_mensagem(string,palavra):
  string_minuscula = string.lower()
  palavra_minuscula = palavra.lower()
  if palavra_minuscula in string_minuscula:
        print(f'Palavra encontrada: {palavra.upper()}')
  else:
        print('Palavra no encontrada!')
entrada1 = "abksjdIntrocompaksdj"
palavra1 = "introcomp"
decifrar_mensagem(entrada1, palavra1)

entrada2 = "abksabksabks"
palavra2 = "introcomp"
decifrar_mensagem(entrada2, palavra2)
````

## Exerccio 6.3.
Escreva um cdigo que receba como entrada dois nmeros $(N, M)$ e, na sada, exiba na tela `N eh multiplo de M` se $N$ for mltiplo de $M$ e, caso contrrio, exiba `N nao eh multiplo de M`.

**Entrada:**\
20\
2\
**Sada:**\
20 eh multiplo de 2\
\
**Entrada:**\
21\
2\
**Sada:**\
21 nao eh multiplo de 2

````python
def verificar_multiplo(N,M):
    if N % M == 0:
        print(f"{N} eh multiplo de {M}")
    else:
        print(f"{N} nao eh multiplo de {M}")

entrada1 = verificar_multiplo(20,2)
entrada2 = verificar_multiplo(21,2)
````

## Exerccio 6.4.
Vamos escrever um programa que calcula a quantia total de dinheiro em uma conta aps uma sequncia de operaes. Teremos 3 operaes:

- D -> Depositar
- S -> Sacar
- E -> Encerrar


| Entrada: | Sada |
| :------- | :----: |
|D<br>300<br>D<br>300<br>S<br>200<br>D<br>100<br>E<br>| 500 |



````python
def calcular_conta():
    saldo = 0.0

    while True:
        print(f"Saldo Atual : R${saldo:.2f}")
        operacao = input("Digite a operao (D - Depositar, S - Sacar, E - Encerrar): ").strip().upper()

        if operacao == 'D':
            valor= float(input("Digite o valor a ser depositado: "))
            saldo = saldo + valor
            print(f"Depsito de R${valor:.2f} realizado com sucesso. Saldo atual R${saldo:.2f}")


        elif operacao == 'S':
          valor = float(input("Digite o valor a ser sacado: "))
          if valor > saldo:
            print("Saldo insuficiente para saque.")
          else:
            saldo -= valor
            print(f"Saque de R${valor:.2f} realizado com sucesso. Saldo atual de R${saldo:.2f} ")


        elif operacao == 'E':
            print(f"Operaes encerradas. Saldo final: R${saldo:.2f}")
            break

        else:
            print("Operao invlida! Por favor, escolha D, S ou E.")

# Chamada da funo
calcular_conta()
````
