Sua principal funo  garantir que determinadas instrues s aconteam caso a condio atribuda ao `if` seja verdadeira.

Voltando ao problema dado no incio da aula (aquele da comparao entre dois valores). Precisamos fazer com  que o computador:

`imprima o valor 'a'` `se` `o valor 'a' for maior que o valor 'b'`

Utilizando o fluxograma da Figura 1, de incio precisamos ter os valores de **a** e de **b**. Para verificar qual  maior, vamos fazer o teste de condio: ```a > b```.

Se a afirmao for `Verdadeira`, ou seja, se `a` for maior do que `b`, podemos imprimir o valor de `a`. Entendeu? Agora vamos ver como escrever coisas desse tipo em um programa de verdade!

Ento... para imprimir um nmero 'a' caso ele seja maior que outro nmero 'b', faramos:



```python
if a > b:
    print(a)
```




Primeiramente, perceba o espaamento que existe  esquerda do comando `print(a)` (comparado com a linha em cima dele). Essa prtica, chamada de **indentao**,  fundamental para que cdigos em Python funcionem corretamente (sendo optativa em outras linguagens de programao). A tecla ideal para fazer este espaamento  o TAB (no seu teclado, fica em cima de Caps Lock), porm tambm pode ser feito com a barra de espao. Lembrando que o TAB cria 4 espaos, ento voce precisa seguir essa regra para que o seu programa funcione.

Com relao  condio colocada logo depois do operador `if`, ela pode ser qualquer valor ou expresso lgica. Quando a condio retornar o valor booleano True (ou um valor diferente de zero), a ao ser executada. Caso a condio retorne o booleano `False`, `None` ou `zero`, a ao no  executada e o programa segue normalmente. O `:` tambm faz parte da sintaxe, ele indica o final da sua condio, apartir dele estaro os comandos, claro respeitando sempre a identao. Vamos testar o exemplo anterior:

````python
# Troque o valor destas variveis e verifique as diferentes sadas de acordo
# com as comparaes realizadas
valor_a = 450.5
valor_b = 472.5

# Teste de condio
if valor_a < valor_b:
  print(valor_a)
````

O mais interessante de se observar neste exemplo  que, quando a expresso lgica do `if`  falsa, o programa continua sendo executado e apenas ignora a ao englobada.

Abaixo temos um exemplo utilizando os operadores de comparao j ensinados em aulas anteriores:

````python
# Troque o valor destas variveis e verifique as diferentes sadas de acordo
# com as comparaes realizadas
meuDinheiro = 2.50
precoCoxinha = 2

if meuDinheiro > precoCoxinha:
    print("Vou comer uma coxinha e ainda vai sobrar dinheiro!")
    troco = meuDinheiro - precoCoxinha
    print("Viu! Sobrou  R$", troco)

if meuDinheiro == precoCoxinha:
    print("Ser que eu devo gastar todo meu dinheiro em uma coxinha?")

print("Com fome ou no, vida que segue")
````

Agora que entendemos como os operadores `if` e `else` funcionam, vamos tentar voltar ao nosso problema inicial. Vamos ento analisar mais uma propriedade de operadores condicionais: o `elif`<br>

O operador `elif`  uma contrao do termo `else if` (se voc j viu linguagens como C talvez conhea dessa forma) e significa `caso contrrio, se...`. Ele  muito til para checar mais de uma condio e criar mais de duas possibilidades de ao.<br>

Voltando para o python, vamos analisar um pouco a sua sintaxe:



```python
if 'Primeira condio':
    'Alguma ao'
elif 'Alguma outra condio':
    'Alguma outra ao'
elif: 'Alguma outra condio, diferente das de cima':
    'Alguma outra ao, diferente das de cima'
else:
    'Alguma outra ao, caso nenhuma das condies sejam atendidas'  
```

Entendeu? Ento vamos explorar essa propriedade: Podemos escrever quantos `elif` quisermos, mas apenas um `if` (Primeira condio) e um `else`(Caso nenhuma condio tenha sido atendida).

Observao: Podemos ter vrios `if` em sequncia como:


```python
if 'Primeira condio':
    'Alguma ao'

if 'Segunda condio':
    'Alguma ao'
```

Nesta situao, os dois `if`s so independentes... isto , a primeira condio ser atendida no exclui a outra de tambm ser! Observe como so `if`s separados e no `if/else`.

Por isso que utilizamos o `elif`, pois com ele a prxima condio s ser verificada se a anterior for `False`.

Vamos fazer um programa que, dada uma quantia de dinheiro na sua carteira, decide se voc vai comprar uma coxinha ou um pastel ou passar fome.

````python
# Aqui vai a quantidade de dinheiro que voce tem na sua carteira.
# Altere o valor para testar o if/elif/else.
carteira = 3

# Aqui so os valores de coxinha ou pastel
coxinha = 4
pastel = 3

# Fica atento a ordem dos if/elif.
if carteira >= coxinha:
    print('Vou comprar uma coxinha!')

    # Paga a coxinha
    carteira = carteira - coxinha

elif carteira >= pastel:
    print('Vou comprar um pastel!')

    # Paga o pastel
    carteira = carteira - pastel

else:
    print('Afs, vou passar fome :(')

print("No final tenho: R$", carteira)
````

# 6. Aninhamento de operadores

Um ltimo tema a ser abordado  o aninhamento de expresses `if`, `else` e `elif`:
Podemos escrever expresses do tipo `if` dentro delas mesmas. Com isso, podemos criar condies dentro de condies, filtrando ainda mais nossos valores. Observe o exemplo abaixo:



```python
if 'Primeira condio':
    if 'Segunda condio':
        'Alguma instruo'
    else:
        'Outra instruo'
```




No exemplo, primeiro verificamos uma condio inicial. Caso essa seja verdadeira, verificamos uma segunda condio. Se a segunda condio for verdadeira, podemos fazer alguma instruo, caso contrrio, podemos fazer uma outra instruo. Vamos ver a Figura 7 para visualizar o fluxo do programa.

<center><b>Figura 7:</b> Estrutura condicional com <i>if/else</i> aninhado.</center>

Neste exemplo usamos um `if/else` aninhado (dentro) de um `if`, porm podemos usar o `elif` tambm, alm de utilizar vrias vezes. Sinta-se  vontade para misturar e utilizar as estruturas condicionais de acordo com sua necessidade... mas cuidado para no se perder no cdigo pois podemos ter vrios nveis de aninhamento!

Para tanto,  importante respeitar a indentao e os dois pontos que seguem essas expresses! Eles ditam quem est dentro de quem, dando sentido ao nosso programa. Vamos avaliar essa propriedade:

````python
# vamos levar em conta trs variveis
a = 1
b = 5
c = 0.5

# mais uma vez, mude as comparaes e os valores e tente acrescentar casos de 'elif'
if a <= b:
    if a > c:
        print('Primeiro "if" aninhado  verdadeiro')
    else:
        print('Primeiro "if" aninhado  falso')
else:
    print('Primeiro "if"  falso')
````

Vamos imaginar uma situao em que voc queira ir para o cinema, mas precisa primeiro verificar se tem dinheiro para a passagem e o filme. Voc tambm s pode assistir se for no final de semana, durante a semana tem
cursinho. Se for o final de semana e voc tiver dinheiro, voc poder ir ao cinema e l verificar se ir poder comprar alguma coisa para comer enquanto assiste um filme.

Cpia de Texto do seu pargrafo.jpg

<center><b>Figura 8:</b> Exemplo visual de alinhamento de operadores.</center>

 Agora, vamos ver isso em cdigo!

````python
# Aqui eh quanto voce tem na carteira. Mude os valores para ver o que acontece
carteira = 60

# Aqui sao os precos
passagem = 8
bilhete = 30
pipoca = 20
refri = 10

# Aqui eh uma variavel booleana que afirma se eh um dia do final de semana ou nao
eh_final_de_semana = True

# Primeiro precisamos verificar se a gente tem na carteira um valor maior que a soma da passagem + bilhete.
# Se nao for maior, significa que nao temos dinheiro suficiente para ir ao cinema ver um filme.
if carteira >= (passagem + bilhete) and eh_final_de_semana:
    print("Vou ao cinema!")

    # Como a gente pode ir ao cinema, precisamos tirar o valor da passagem e do ingresso da carteira.
    carteira = carteira - passagem - bilhete

    # Primeiro vamos ver se podemos comprar pipoca + refri. Basta ver se a gente tem mais dinheiro na carteira.
    if carteira >= (pipoca + refri):
        print("Irei comer pipoca com refri :D")

    # Se nao poder comprar os dois, podemos verificar se temos pelo menos dinheiro para comprar so a pipoca ou so o refri.
    elif carteira >= pipoca or carteira >= refri:
        print("No consigo comprar os dois. Vou comprar s um :/")

    # Se nao tem dinheiro suficiente para comprar pipoca ou refri, ficamos sem comida
    else:
        print("Cinema  pra ver filme. No preciso de comida :|")

# Se nao for o final de semana, nao podemos ir
elif not eh_final_de_semana:
    print("Hoje no  final de semana, tenho que estudar. No vou poder ir ao cinema :(")

# Senao entrou no elif anterior, significa que eh final de semana, mas a gente nao tem dinheiro suficiente
else:
    print("No tenho dinheiro suficiente. No vou poder ir ao cinema :(")
````

## 7.2. Adicionando a quantidade de chutes


Agora conseguimos apresentar uma mensagem dependendo do nosso contexto. O nosso jogo ainda parece incompleto, no jogo da forca podemos advinhar vrias vezes, mas por enquanto o jogo nos permite chutar apenas uma vez! Queremos poder chutar palavras diferentes vrias vezes, e para isso precisamos definir quantas vezes o usurio poder tentar acertar a palavra.

No nosso jogo, vamos determinar quantas vezes o usurio pode chutar alguma palavra a partir da escolha da dificuldade do jogo. A dificuldade do nosso jogo vai influenciar na quantidade mxima de chutes possvel.

Teremos 10 nveis no total. No nvel 1, o jogador pode chutar 10 vezes e no nvel 10, pode chutar apenas 1 vez. Cada nvel torna o jogo mais difcil, tirando 1 chance dele. Ento, no nvel 2, podem ser feitas 9 tentativas, 8 no nvel 3 e assim por adiante.

Podemos ento perceber um padro no nosso cdigo:
```
Se o jogador tiver no nvel 1, ele tem 10 tentativas.
Se o jogador tiver no nvel 2, ele tem 9 tentativas.
...
Se o jogador tiver no nvel 9, ele tem 2 tentativas.
Se o jogador tiver no nvel 10, ele tem 1 tentativa.
```

````python
dificuldade = int(input("Escolha a dificuldade (de 1 a 10): "))

# Definindo o nmero de tentativas com base na dificuldade:
tentativas_possiveis = 11 - dificuldade
````

# 8. Mini-Projeto II: Calcule seu imposto de renda com **if-else** e **dicionrios**

IMPORTANTE: O uso de dicionrios ser melhor explicado em outra aula, mas, tente adiantar-se para aprender mais! Voltaremos no assunto muito em breve. [Documentao](https://docs.python.org/pt-br/3/tutorial/datastructures.html#dictionaries).

**Tabela de IRRF**:

Conhecendo a base real do imposto, pode-se identificar a faixa de percentual em que o funcionrio se encaixa. As faixas atualizadas so:

* at R\$ 1.903,98: iseno;

* 1 faixa: 7,5\% para bases de R\$ 1.903,99 a R\$ 2.826,65;

* 2 faixa: 15\% para bases de R\$ 2.826,66 a R\$ 3.751,05;

* 3 faixa: 22,5\% para bases de R\$ 3.751,06 a R\$ 4.664,68;

* 4 faixa: 27,5\% para bases a partir de R\$ 4.664,69.


Aps verificado o imposto de acordo com a alquota da sua base de clculo,  deduzido um valor padro, o que permite obter o valor final do imposto. Cada faixa de tributao tem uma deduo definida, da seguinte forma:

* 1 faixa: R\$ 142,80;

* 2 faixa: R\$ 354,80;

* 3 faixa: R\$ 636,13;

* 4 faixa: R\$ 869,36.

Abaixo  mostrada a contabilizao do imposto no caso de um exemplo:

* R\$ 2.250,41 x 7,5\% = R\$ 168,78;
* R\$ 168,78  R\$ 142,80 = R\$ 25,98 de IR a ser retido na fonte.

Crie um programa que dado um salrio bruto digitado fornece:

* Salrio bruto;

* Faixa de IR (em percentual);

* Imposto de renda a pagar (antes do desconto do valor padro);

* Valor padro (de acordo com a faixa de IR);

* Imposto retido na fonte;

* Salrio lquido.

````python
salario = input('Digite o salrio bruto: ')
salario = float(salario)

# Dicionrio faixa -> percentual de desconto.
fDesc = {0: 0.0, 1: 7.5, 2: 15.0, 3: 22.5, 4: 27.5}

# Dicionrio faixa -> valor padro.
fVPad = {0: 0, 1: 142.80, 2: 354.80, 3: 636.13, 4: 869.36}

# Verificando em qual faixa de imposto se enquadra o salrio bruto.
if (salario <= 1903.98):
  faixa = 0
elif (salario >= 1903.99)and(salario <= 2826.65):
  faixa = 1
elif (salario >= 2826.66)and(salario <= 3751.05):
  faixa = 2
elif (salario >= 3751.06)and(salario <= 4664.68):
  faixa = 3
else:   # implcito salario >= 4664.69
  faixa = 4

imposto = salario*(fDesc[faixa]/100)
vpad    = fVPad[faixa]
impagar = imposto - vpad
liquido = salario - impagar

mensagem1 = "Bruto: %10.2f, Faixa: %d, Imposto: %5.2f" % (salario, faixa, imposto)
mensagem2 = "Padro: %5.2f, Pagar: %5.2f, Liquido: %5.2f" % (vpad, impagar, liquido)
print(mensagem1)
print(mensagem2)

````

# 9. O que pode dar errado?


No decorrer da aula, voc possivelmente teve alguns problemas enquanto programava, e para tentar solucionar as dvidas mais frequentes temos esta seo.

Um dos erros mais frequentes no uso das condicionais em Python  a falta da indentao (IndentationError). Observe o cdigo abaixo:

````python
a = True

if a:
  print("123")
````

Deu problema, n? Como dito durante a aula, a indentao (o espao do comando em relao ao lado esquerdo)  fundamental para a interpretao correta do algoritmo. Caso tenha se esquecido, para dar este espaamento basta apertar a tecla TAB.

O smbolo de dois pontos aps a condio tambm  fundamental. Se voc esquecer de utiliz-los, o computador vai indicar um erro.

````python
a = True

if a:
    print("123")
````

````python
a = True

if a:
    print("Verdade")
else:
    print("Mentira")
````

Fique atento tambm a ordem dos `if/elif/else`. Podemos utilizar um if isolado como no exemplo abaixo:



```
if True:
    print("isso  verdade")
```

Mas nunca podemos iniciar nossa sequncia com um `elif` ou um `else`. Observe as duas clulas de cdigo abaixo. Ambas iro sinalizar erro de sintxe.


````python
else:
    print("Fui executado?")
````

````python
nota = 10
elif a >= 7:
    print("Estou acima da mdia")
````

Alm disso, o `else` significa caso contrrio, portanto ele no precisaria verificar algum tipo de condio. Por exemplo:

````python
chover = False

if chover:
    print("Preciso de abrir meu guarda chuva")
else:
    print("Fez sol")
````

O exemplo acima est errado, pois o `else` est sendo acompanhado por uma condio. Para corrigir este cdigo devemos utilizar o `else` da seguinte forma:

````python
chover = True

if chover:
    print("Preciso de abrir meu guarda chuva")
else:
    print("Fez sol")
````

Tambm fique atento  ordem dos operadores. Caso for utilizar `elif` e `else`, o `else` nunca pode vir antes do `elif`.

````python
nota = 10

if nota >=7:
    print("Passou :)")
elif nota >= 5:
    print("Voc estar de recuperao :|")
else:
    print("Reprovou :(")
````
