# 3. Operador if

O primeiro operador condicional que temos é o `if`, que pode ser traduzido para `se`.

Sua principal função é garantir que determinadas instruções só aconteçam caso a condição atribuída ao `if` seja verdadeira.

```python
if 'Alguma condição':
    'Alguma ação'
```

Vamos tentar entender melhor com o auxílio do fluxograma a abaixo (Figura 1):<br>



<center></center>

<center><b>Figura 1:</b> Estrutura condicional simples, representada em um fluxograma.</center>


Para ler o fluxograma acima, é preciso entender o que cada forma da figura faz.
* O **retângulo laranja escuro** representa o início do nosso fluxo (início do código ou de uma seção dele) e os dados de entrada para o programa.
* O **círculo pequeno** representa um estado do programa, ou seja, como se encontrava o programa naquele instante. Ele é utilizado para sinalizar que vamos de um estado para outro, através de diferentes caminhos.
* O **losango** representa uma condição. Essa condição pode ter dois resultados, verdadeiro ou falso. Vamos utilizá-lo bastante nas próximas figuras!
* O **retângulo laranja claro** representa uma instrução (ou um conjunto de instruções), representando uma ação que deve ser feita.

Imagine que você está em um restaurante e olha para o cardápio para decidir o que pedir. Se você escolher um prato que está disponível, o garçom irá anotar o seu pedido e levá-lo para a cozinha. Esta é uma forma simples de compreender como o `if` funciona: Se você cumprir a condição (no caso da Figura 2, escolher o pedido), a instrução será realizada (no caso da Figura 2, o pedido será realizado).

Voltando ao problema dado no início da aula (aquele da comparação entre dois valores). Precisamos fazer com  que o computador:

`imprima o valor 'a'` `se` `o valor 'a' for maior que o valor 'b'`

Utilizando o fluxograma da Figura 1, de início precisamos ter os valores de **a** e de **b**. Para verificar qual é maior, vamos fazer o teste de condição: ```a > b```.

Se a afirmação for `Verdadeira`, ou seja, se `a` for maior do que `b`, podemos imprimir o valor de `a`. Entendeu? Agora vamos ver como escrever coisas desse tipo em um programa de verdade!

Então... para imprimir um número 'a' caso ele seja maior que outro número 'b', faríamos:



```python
if a > b:
    print(a)
```




Primeiramente, perceba o espaçamento que existe à esquerda do comando `print(a)` (comparado com a linha em cima dele). Essa prática, chamada de **indentação**, é fundamental para que códigos em Python funcionem corretamente (sendo optativa em outras linguagens de programação). A tecla ideal para fazer este espaçamento é o TAB (no seu teclado, fica em cima de Caps Lock), porém também pode ser feito com a barra de espaço. Lembrando que o TAB cria 4 espaços, então voce precisa seguir essa regra para que o seu programa funcione.

Com relação à condição colocada logo depois do operador `if`, ela pode ser qualquer valor ou expressão lógica. Quando a condição retornar o valor booleano True (ou um valor diferente de zero), a ação será executada. Caso a condição retorne o booleano `False`, `None` ou `zero`, a ação não é executada e o programa segue normalmente. O `:` também faz parte da sintaxe, ele indica o final da sua condição, apartir dele estarão os comandos, claro respeitando sempre a identação. Vamos testar o exemplo anterior:

````python
# Troque o valor destas variáveis e verifique as diferentes saídas de acordo
# com as comparações realizadas
valor_a = 450.5
valor_b = 472.5

# Teste de condição
if valor_a < valor_b:
  print(valor_a)
````

O mais interessante de se observar neste exemplo é que, quando a expressão lógica do `if` é falsa, o programa continua sendo executado e apenas ignora a ação englobada.

Abaixo temos um exemplo utilizando os operadores de comparação já ensinados em aulas anteriores:

````python
# Troque o valor destas variáveis e verifique as diferentes saídas de acordo
# com as comparações realizadas
meuDinheiro = 2.50
precoCoxinha = 2

if meuDinheiro > precoCoxinha:
    print("Vou comer uma coxinha e ainda vai sobrar dinheiro!")
    troco = meuDinheiro - precoCoxinha
    print("Viu! Sobrou  R$", troco)

if meuDinheiro == precoCoxinha:
    print("Será que eu devo gastar todo meu dinheiro em uma coxinha?")

print("Com fome ou não, vida que segue")
````

# 5. Operador elif

Agora que entendemos como os operadores `if` e `else` funcionam, vamos tentar voltar ao nosso problema inicial. Vamos então analisar mais uma propriedade de operadores condicionais: o `elif`<br>

O operador `elif` é uma contração do termo `else if` (se você já viu linguagens como C talvez conheça dessa forma) e significa `caso contrário, se...`. Ele é muito útil para checar mais de uma condição e criar mais de duas possibilidades de ação.<br>



Na figura 5 temos um exemplo `MUITO INEFICIENTE` de utilização do `elif` para saber se um número é par:

image.png

<center><b>Figura 5:</b> Exemplo de mau uso do operador <i>elif</i>.</center>

Voltando para o python, vamos analisar um pouco a sua sintaxe:



```python
if 'Primeira condição':
    'Alguma ação'
elif 'Alguma outra condição':
    'Alguma outra ação'
elif: 'Alguma outra condição, diferente das de cima':
    'Alguma outra ação, diferente das de cima'
else:
    'Alguma outra ação, caso nenhuma das condições sejam atendidas'  
```

Entendeu? Então vamos explorar essa propriedade: Podemos escrever quantos `elif` quisermos, mas apenas um `if` (Primeira condição) e um `else`(Caso nenhuma condição tenha sido atendida).

Observação: Podemos ter vários `if` em sequência como:


```python
if 'Primeira condição':
    'Alguma ação'

if 'Segunda condição':
    'Alguma ação'
```

Nesta situação, os dois `if`s são independentes... isto é, a primeira condição ser atendida não exclui a outra de também ser! Observe como são `if`s separados e não `if/else`.

Por isso que utilizamos o `elif`, pois com ele a próxima condição só será verificada se a anterior for `False`.

Podemos visualizar melhor o que está acontecendo dentro do programa quando utlizamos `elif`com ajuda do fluxograma (Figura 6).

<center><b>Figura 6:</b> Estrutura condicional com <i>else</i> e <i>elif</i>.</center>

Vamos fazer um programa que, dada uma quantia de dinheiro na sua carteira, decide se você vai comprar uma coxinha ou um pastel ou passar fome.

````python
# Aqui vai a quantidade de dinheiro que voce tem na sua carteira.
# Altere o valor para testar o if/elif/else.
carteira = 3

# Aqui são os valores de coxinha ou pastel
coxinha = 4
pastel = 3

# Fica atento a ordem dos if/elif.
if carteira >= coxinha:
    print('Vou comprar uma coxinha!')

    # Paga a coxinha
    carteira = carteira - coxinha

elif carteira >= pastel:
    print('Vou comprar um pastel!')

    # Paga o pastel
    carteira = carteira - pastel

else:
    print('Afs, vou passar fome :(')

print("No final tenho: R$", carteira)
````

A ordem é muito importante! O que acontece no código de cima se pastel tem o mesmo valor de coxinha? Ele entra no elif? Por quê?

# 6. Aninhamento de operadores

Um último tema a ser abordado é o aninhamento de expressões `if`, `else` e `elif`:
Podemos escrever expressões do tipo `if` dentro delas mesmas. Com isso, podemos criar condições dentro de condições, filtrando ainda mais nossos valores. Observe o exemplo abaixo:



```python
if 'Primeira condição':
    if 'Segunda condição':
        'Alguma instrução'
    else:
        'Outra instrução'
```




No exemplo, primeiro verificamos uma condição inicial. Caso essa seja verdadeira, verificamos uma segunda condição. Se a segunda condição for verdadeira, podemos fazer alguma instrução, caso contrário, podemos fazer uma outra instrução. Vamos ver a Figura 7 para visualizar o fluxo do programa.

<center><b>Figura 7:</b> Estrutura condicional com <i>if/else</i> aninhado.</center>

Neste exemplo usamos um `if/else` aninhado (dentro) de um `if`, porém podemos usar o `elif` também, além de utilizar várias vezes. Sinta-se à vontade para misturar e utilizar as estruturas condicionais de acordo com sua necessidade... mas cuidado para não se perder no código pois podemos ter vários níveis de aninhamento!

Para tanto, é importante respeitar a indentação e os dois pontos que seguem essas expressões! Eles ditam quem está dentro de quem, dando sentido ao nosso programa. Vamos avaliar essa propriedade:

````python
# vamos levar em conta três variáveis
a = 1
b = 5
c = 0.5

# mais uma vez, mude as comparações e os valores e tente acrescentar casos de 'elif'
if a <= b:
    if a > c:
        print('Primeiro "if" aninhado é verdadeiro')
    else:
        print('Primeiro "if" aninhado é falso')
else:
    print('Primeiro "if" é falso')
````

Vamos imaginar uma situação em que você queira ir para o cinema, mas precisa primeiro verificar se tem dinheiro para a passagem e o filme. Você também só pode assistir se for no final de semana, durante a semana tem
cursinho. Se for o final de semana e você tiver dinheiro, você poderá ir ao cinema e lá verificar se irá poder comprar alguma coisa para comer enquanto assiste um filme.

Cópia de Texto do seu parágrafo.jpg

<center><b>Figura 8:</b> Exemplo visual de alinhamento de operadores.</center>

 Agora, vamos ver isso em código!

````python
# Aqui eh quanto voce tem na carteira. Mude os valores para ver o que acontece
carteira = 60

# Aqui sao os precos
passagem = 8
bilhete = 30
pipoca = 20
refri = 10

# Aqui eh uma variavel booleana que afirma se eh um dia do final de semana ou nao
eh_final_de_semana = True

# Primeiro precisamos verificar se a gente tem na carteira um valor maior que a soma da passagem + bilhete.
# Se nao for maior, significa que nao temos dinheiro suficiente para ir ao cinema ver um filme.
if carteira >= (passagem + bilhete) and eh_final_de_semana:
    print("Vou ao cinema!")

    # Como a gente pode ir ao cinema, precisamos tirar o valor da passagem e do ingresso da carteira.
    carteira = carteira - passagem - bilhete

    # Primeiro vamos ver se podemos comprar pipoca + refri. Basta ver se a gente tem mais dinheiro na carteira.
    if carteira >= (pipoca + refri):
        print("Irei comer pipoca com refri :D")

    # Se nao poder comprar os dois, podemos verificar se temos pelo menos dinheiro para comprar so a pipoca ou so o refri.
    elif carteira >= pipoca or carteira >= refri:
        print("Não consigo comprar os dois. Vou comprar só um :/")

    # Se nao tem dinheiro suficiente para comprar pipoca ou refri, ficamos sem comida
    else:
        print("Cinema é pra ver filme. Não preciso de comida :|")

# Se nao for o final de semana, nao podemos ir
elif not eh_final_de_semana:
    print("Hoje não é final de semana, tenho que estudar. Não vou poder ir ao cinema :(")

# Senao entrou no elif anterior, significa que eh final de semana, mas a gente nao tem dinheiro suficiente
else:
    print("Não tenho dinheiro suficiente. Não vou poder ir ao cinema :(")
````

Agora, tente realizar mais um pequeno projeto. Desta vez, imagine que você tenha um site e precise avaliar se alguns usuários podem ou não criar uma conta para o seu serviço. Eles te fornecem um nome, uma senha e a idade deles, e você deve dizer se esses dados são válidos ou não, considerando que:
    
    -O nome de usuário deve ter pelo menos 4 letras;
    -A senha deve ter ao menos 8 caracteres;
    -A idade deve ser de, no mínimo, 15 anos.

Este exercício tem várias formas de resolver, podendo ser utilizado if aninhados ou operações lógicas.

````python
# esses são os dados do usuário. Lembre-se de testar dados diferentes!
nome = 'jao'
senha = 'senha123'
idade = 18

#agora escreva a parte que deve avaliar se o usuário é válido ou não
def validar_usuario(nome,senha,idade):
  validar_nome = False
  validar_senha = False
  validar_idade = False
  if len(nome)>=4:
    validar_nome=True
  if len(senha)>=8:
    validar_senha=True
  if idade >15:
    validar_idade=True
  return validar_nome, validar_senha, validar_idade
validar_nome,validar_senha,validar_idade = validar_usuario(nome,senha,idade)

if validar_nome and validar_senha and validar_idade:
  print('usuário valido!')
else:
  print('usuário invalido')
````

Na aula passada, vimos como podemos validar se o jogador acertou a palavra, após um chute (não uma letra). Vamos relembrar o código da aula passada.

````python
# Perguntando a palavra para o usuário:
palavra = input("Qual a palavra a ser decifrada? ")

# Tratando a palavra:
palavra = palavra.strip().lower()

# Perguntando o chute do usuário:
chute = input("Chute uma palavra: ")

# Tratando o chute:
chute = chute.strip().lower()

# Verificando se o usuário acertou o chute:
acertou = palavra == chute

print(f"O usuário acertou? {acertou}")
````

## 7.1. Adicionando mensagem de acerto

Observar que o `print` não fala explicitamente se o usuário acertou ou não, apenas mostra o valor da variável `acertou`. Nessa aula, aprendemos como seguir determinadas instruções dependendo de alguma situação, como o valor de uma variável, utilizando a estrutura `if`. Vamos então modificar o `print` do código da aula passada para apenas apresentar "Você acertou!" caso o chute estiver correto e "Chute incorreto!" se o chute estiver errado.

````python
# Perguntando a palavra para o usuário:
palavra = input("Qual a palavra a ser decifrada? ")

# Tratando a palavra:
palavra = palavra.strip().lower()

# Perguntando o chute do usuário:
chute = input("Chute uma palavra: ")

# Tratando o chute:
chute = chute.strip().lower()

# Verificando se o usuário acertou o chute:
acertou = palavra == chute

if acertou:
    print("Você acertou!")
else:
    print("Chute incorreto!")
````

## 7.2. Adicionando a quantidade de chutes


Agora conseguimos apresentar uma mensagem dependendo do nosso contexto. O nosso jogo ainda parece incompleto, no jogo da forca podemos advinhar várias vezes, mas por enquanto o jogo nos permite chutar apenas uma vez! Queremos poder chutar palavras diferentes várias vezes, e para isso precisamos definir quantas vezes o usuário poderá tentar acertar a palavra.

No nosso jogo, vamos determinar quantas vezes o usuário pode chutar alguma palavra a partir da escolha da dificuldade do jogo. A dificuldade do nosso jogo vai influenciar na quantidade máxima de chutes possível.

Teremos 10 níveis no total. No nível 1, o jogador pode chutar 10 vezes e no nível 10, pode chutar apenas 1 vez. Cada nível torna o jogo mais difícil, tirando 1 chance dele. Então, no nível 2, podem ser feitas 9 tentativas, 8 no nível 3 e assim por adiante.

Podemos então perceber um padrão no nosso código:
```
Se o jogador tiver no nível 1, ele tem 10 tentativas.
Se o jogador tiver no nível 2, ele tem 9 tentativas.
...
Se o jogador tiver no nível 9, ele tem 2 tentativas.
Se o jogador tiver no nível 10, ele tem 1 tentativa.
```

Observe que a soma de nível e tentativas sempre é igual a 11. Podemos utilizar essa relação para criar uma fórmula para representar quantas tentativas serão possíveis. O nível é o valor guardado pela variável `dificuldade`. Se sabemos que:

```
11 = dificuldade + tentativas
```
Logo:

```
tentativas = 11 - dificuldades
```

Vamos implementar isso em Python!



````python
dificuldade = int(input("Escolha a dificuldade (de 1 a 10): "))

# Definindo o número de tentativas com base na dificuldade:
tentativas_possiveis = 11 - dificuldade
````

Mas o que acontece se o jogador colocar um número maior que 10, ou menor 1? Temos que verificar os limites! O valor de dificuldade é maior que 10 ou menor que 1? Então temos que finalizar o jogo, pois apenas aceitamos valores de 1 a 10. Podemos fazer isso usando o `if`. O nosso número não pode ser maior que 10 **ou** menor que 1. Preste atenção ao **ou**, ele indica que basta uma dessas situações ser verdadeira. Vamos traduzir isso para código?

````python
dificuldade = int(input("Escolha a dificuldade (de 1 a 10): "))

# Verificando se a dificuldade está correta:
if dificuldade > 10 or dificuldade < 1:
    quit()
````

Observe que aqui utilizamos uma função nova: `quit()`. Utilizamos ela quando algum erro é detectado e precisamos terminar a execução do programa imediatamente... antes do restante do código ser rodado.

Adicionando para o nosso código anterior temos então:

````python
dificuldade = int(input("Escolha a dificuldade (de 1 a 10): "))

# Verificando se a dificuldade está correta:
if dificuldade > 10 or dificuldade < 1:
    quit()

# Definindo o número de tentativas com base na dificuldade:
tentativas_possiveis = 11 - dificuldade

# Perguntando a palavra para o usuário:
palavra = input("Qual a palavra a ser decifrada? ")

# Tratando a palavra:
palavra = palavra.strip().lower()

# Verificando se o usuário acertou o chute:
acertou = palavra == chute

if acertou:
    print("Você acertou!")
else:
    print("Chute incorreto!")
````

# 8. Mini-Projeto II: Calcule seu imposto de renda com **if-else** e **dicionários**

Fonte: https://blog.convenia.com.br/como-calcular-irrf-na-folha-de-pagamento/

IMPORTANTE: O uso de dicionários será melhor explicado em outra aula, mas, tente adiantar-se para aprender mais! Voltaremos no assunto muito em breve. [Documentação](https://docs.python.org/pt-br/3/tutorial/datastructures.html#dictionaries).

**Tabela de IRRF**:

Conhecendo a base real do imposto, pode-se identificar a faixa de percentual em que o funcionário se encaixa. As faixas atualizadas são:

* até R\$ 1.903,98: isenção;

* 1ª faixa: 7,5\% para bases de R\$ 1.903,99 a R\$ 2.826,65;

* 2ª faixa: 15\% para bases de R\$ 2.826,66 a R\$ 3.751,05;

* 3ª faixa: 22,5\% para bases de R\$ 3.751,06 a R\$ 4.664,68;

* 4ª faixa: 27,5\% para bases a partir de R\$ 4.664,69.


Após verificado o imposto de acordo com a alíquota da sua base de cálculo, é deduzido um valor padrão, o que permite obter o valor final do imposto. Cada faixa de tributação tem uma dedução definida, da seguinte forma:

* 1ª faixa: R\$ 142,80;

* 2ª faixa: R\$ 354,80;

* 3ª faixa: R\$ 636,13;

* 4ª faixa: R\$ 869,36.

Abaixo é mostrada a contabilização do imposto no caso de um exemplo:

* R\$ 2.250,41 x 7,5\% = R\$ 168,78;
* R\$ 168,78 – R\$ 142,80 = R\$ 25,98 de IR a ser retido na fonte.

Crie um programa que dado um salário bruto digitado fornece:

* Salário bruto;

* Faixa de IR (em percentual);

* Imposto de renda a pagar (antes do desconto do valor padrão);

* Valor padrão (de acordo com a faixa de IR);

* Imposto retido na fonte;

* Salário líquido.

````python
salario = input('Digite o salário bruto: ')
salario = float(salario)

# Dicionário faixa -> percentual de desconto.
fDesc = {0: 0.0, 1: 7.5, 2: 15.0, 3: 22.5, 4: 27.5}

# Dicionário faixa -> valor padrão.
fVPad = {0: 0, 1: 142.80, 2: 354.80, 3: 636.13, 4: 869.36}

# Verificando em qual faixa de imposto se enquadra o salário bruto.
if (salario <= 1903.98):
  faixa = 0
elif (salario >= 1903.99)and(salario <= 2826.65):
  faixa = 1
elif (salario >= 2826.66)and(salario <= 3751.05):
  faixa = 2
elif (salario >= 3751.06)and(salario <= 4664.68):
  faixa = 3
else:   # implícito salario >= 4664.69
  faixa = 4

imposto = salario*(fDesc[faixa]/100)
vpad    = fVPad[faixa]
impagar = imposto - vpad
liquido = salario - impagar

mensagem1 = "Bruto: %10.2f, Faixa: %d, Imposto: %5.2f" % (salario, faixa, imposto)
mensagem2 = "Padrão: %5.2f, Pagar: %5.2f, Liquido: %5.2f" % (vpad, impagar, liquido)
print(mensagem1)
print(mensagem2)

````

# 9. O que pode dar errado?


No decorrer da aula, você possivelmente teve alguns problemas enquanto programava, e para tentar solucionar as dúvidas mais frequentes temos esta seção.

Um dos erros mais frequentes no uso das condicionais em Python é a falta da indentação (IndentationError). Observe o código abaixo:

````python
a = True

if a:
  print("123")
````

Deu problema, né? Como dito durante a aula, a indentação (o espaço do comando em relação ao lado esquerdo) é fundamental para a interpretação correta do algoritmo. Caso tenha se esquecido, para dar este espaçamento basta apertar a tecla TAB.

O símbolo de dois pontos após a condição também é fundamental. Se você esquecer de utilizá-los, o computador vai indicar um erro.

````python
a = True

if a:
    print("123")
````

````python
a = True

if a:
    print("Verdade")
else:
    print("Mentira")
````

Fique atento também a ordem dos `if/elif/else`. Podemos utilizar um if isolado como no exemplo abaixo:



```
if True:
    print("isso é verdade")
```

Mas nunca podemos iniciar nossa sequência com um `elif` ou um `else`. Observe as duas células de código abaixo. Ambas irão sinalizar erro de sintáxe.


````python
else:
    print("Fui executado?")
````

````python
nota = 10
elif a >= 7:
    print("Estou acima da média")
````

Além disso, o `else` significa caso contrário, portanto ele não precisaria verificar algum tipo de condição. Por exemplo:

````python
chover = False

if chover:
    print("Preciso de abrir meu guarda chuva")
else:
    print("Fez sol")
````

O exemplo acima está errado, pois o `else` está sendo acompanhado por uma condição. Para corrigir este código devemos utilizar o `else` da seguinte forma:

````python
chover = True

if chover:
    print("Preciso de abrir meu guarda chuva")
else:
    print("Fez sol")
````

Também fique atento à ordem dos operadores. Caso for utilizar `elif` e `else`, o `else` nunca pode vir antes do `elif`.

````python
nota = 10

if nota >=7:
    print("Passou :)")
elif nota >= 5:
    print("Você estar de recuperação :|")
else:
    print("Reprovou :(")
````
