# 2. O que é um programa?



Um programa é uma sequência de instruções que especificam como o computador deve realizar uma tarefa. Essa tarefa pode ser algo como encontrar as raízes de uma equação de segundo grau,  uma busca num texto para encontrar e substituir uma palavra, etc.  

Diferentemente de um algoritmo, um programa é escrito em alguma linguagem formal, chamada de **linguagem de programação**.
Da mesma forma como as pessoas se comunicam umas com as outras por meio de alguma linguagem (português, inglês, francês, etc.), precisamos de um jeito para nos comunicar com o computador. Para isso, utilizamos alguma linguagem de programação, como C, C++, Java, PHP, Lua, etc.

Neste curso, vamos aprender a linguagem de programação chamada **Python**.

Normalmente, as instruções que compõem um programa variam de acordo com a linguagem de programação que você está utilizando, mas é comum em praticamente todos os programas algumas instruções básicas:

- Entrada: receber informações, seja do teclado, de um arquivo, do mouse ou de um sensor de movimento numa câmera, por exemplo;
- Saída: enviar informações, seja no terminal (caso não saiba o que é isso, fique tranquilo! Logo veremos mais sobre), em um arquivo, ou num display;
- Operações Aritméticas: realizar operações matemáticas (soma, subtração, divisão e etc), essencial para a maioria dos programas;
- Condicionais: checar condições e realizar instruções apropriadas;
- Repetições: realizar uma sequência de comandos de forma repetitiva.

Acredite, qualquer programa que você fizer, por mais complicado que seja, é feito de instruções que se aproximam dessas 5 listadas acima.

# 5. Variáveis



Variáveis são espaços na memória, destinados a armazenar informações que serão utilizadas no futuro pelo computador. Pense como uma caixa que armazena algum determinado dado, e esta caixa recebe um nome (uma "etiqueta").

As variáveis podem ter qualquer nome, respeitando algumas regras:

    1. O primeiro caracter precisa ser uma letra (A-Z, a-z) ou um underscore ( _ )
    2. Os demais caracteres podem ser letras, números ou underscore

Há algumas convenções sobre como devemos nomear variáveis. Uma convenção possível sugere que todo o nome da variável deve ser escrito em letras minúsculas e cada palavra deve ser separada por um *underscore* ("_"). Alguns exemplos estão presentes na tabela abaixo:

Nome modelo | Variável seguindo essa convenção
------------|-------------------------
Dinheiro na conta|dinheiro_na_conta
Idade mínima|idade_minima
CPF do Cliente|cpf_cliente

**Deve-se evitar acentos e outros caracteres especiais ao criar variáveis!**

**Dica:** tente atribuir nomes intuitivos e de fácil entendimento para as suas variáveis! Isso facilita a construção do seu código, bem como a sua futura revisão.

Além disso, existem algumas palavras reservadas na linguagem, que não podem nomear variáveis (ainda que respeitem as regras acima). Algumas dessas palavras são: True, False, continue, and, None, etc.

Vejamos algumas das principais funções para manipulacao de strings!

### 8.1. Função `lower`



Transforma todos os caracteres que estão em maiúsculo para minúsculo ( A -> a )

````python
frase = "ISSO É UMA STRING"

frase_modificada = frase.lower();

print(frase_modificada)
````

### 8.2. Função `upper`



Transforma todos os caracteres que estão em minúsculo para maiúsculo ( a -> A )

````python
frase = "está tudo pequeno"

frase_modificada = frase.upper();

print(frase_modificada)
````

### 8.3. Função `strip`

Essa função "ajeita" nossa string e é bem interessante para tratar entradas desse tipo de dado, uma vez que a chance de existirem caracteres especiais a mais diminui bastante.

Vejamos um exemplo sem `strip`.

````python
palavra = " IA-CD  "
palavra == "IA-CD"
````

Agora, veja como fica com o uso de `strip`:

````python
palavra = " IA-CD  ".strip()
palavra == "IA-CD"
````

Você deve ter percebido que `strip` retirou os espaços em branco das extremidades da string.

### 8.4. _f-strings_

Além de tratar nossas _strings_ com `upper`, `lower` e `strip`, existe uma maneira de utilizar uma _string_ como base para uma frase. Esse recurso foi adicionado na versão 3.6 de Python e é extremamente útil.

Veja um exemplo:

````python
nome = input("Qual seu nome? ").strip()
saudacao = f"Bom dia, {nome}!"

print(saudacao)
````

Utilizamos a função input() para podermos obter o que foi escrito pelo usuário. Perceba que o trecho onde existia um `{nome}` foi substituído pelo conteúdo da variável nome. É exatamente assim que utilizamos _f-strings_, basta inserir algo do tipo `{minha_variavel}` no meio de uma _string_. Uma outra característica das _f-strings_ é a necessidade da letra `f` antes do primeiro `"`.

### 8.5. Função `format`

Por fim, vamos aprender como inserir variáveis no meio de uma string (sequência de letras e espaços) sem precisar separá-la com vírgulas ou operadores "+". Essa técnica também é util para manter o código mais legível! Para colocarmos variáveis no meio de uma string, basta colocar `{}` no exato local onde você quer que a variável apareça, e após terminar a string, escreva `.format(nome_da_variável)`. Um exemplo deve tornar as coisas mais simples:<br>

````python
imc = 22.54656
print("Uma pessoa com o IMC {} está dentro do seu peso ideal!".format(imc))
````

Achou esquisito? Geralmente, quando vamos apresentar números com ponto flutuante como esse (chamados de **float** em Python), costumamos arredondá-los para um número menor de casas decimais usando o especificador **.2f**, como no exemplo a seguir:

````python
imc = 22.54656
print("Uma pessoa com o IMC {:.2f} está dentro do seu peso ideal!".format(imc))
````

Podemos utilizar essa formatação também para imprimir strings:

````python
almoco = "strogonoff de frango"
````

````python
#é bem fácil repetir o print que fizemos algumas células acima agora
print("quero {} no almoço!".format(almoco))
````

E se for preciso colocar mais de uma variável na mesma string? Bom, a sintaxe mudará um pouco, mas é bem similar à anterior. <br>
Como sempre, um exemplo deve ilustrar melhor como fazer:<br>

````python
a = "farinha"
b = "ovos"
c = "açúcar"
d = "banana"
tempo = 40

print("Para fazer um bolo de banana, precisamos de {}, {}, {} e {}. Ele fica pronto em {} minutos!".format(a, b, c, d, tempo))
````

## 8.6. Métodos de string

Strings podem ser transformadas usando **métodos**, que são funções que envolvem uma string existente e alguns outros argumentos. Um exemplo é o método `replace`, que substitui todas as instâncias de alguma parte de uma string por alguma alternativa.

Um método é invocado em uma string colocando um `.` após o valor da string, depois o nome do método e, finalmente, parênteses contendo os argumentos. Aqui está um esboço, onde os símbolos `<` e `>` não fazem parte da sintaxe; eles apenas marcam os limites das subexpressões.

> `<string>.<nome do método>(<argumento>, <argumento>, ...)`

Tente prever o resultado desses exemplos e execute-os.

````python
# Substituíndo uma letra
'Hello'.replace('o', 'a')
````

````python
# Substituíndo uma sequência de letras por outra (note que a sequência se repete na palavra)
'hitchhiker'.replace('hi', 'ma')
````

Depois que uma variável é vinculada a um valor de string, os métodos também podem ser invocados nessa variável (lembre-se, "variável" é outra palavra para "nome"). A variável `sharp` não muda neste caso, então uma nova variável é necessária para capturar o resultado.

````python
sharp = 'edged'
hot = sharp.replace('ed', 'ma')
print('sharp:', sharp)
print('hot:', hot)
````

````python
# Chamando duas vezes a função replace
'train'.replace('t', 'ing').replace('in', 'de')
````

A maior parte da programação envolve um trabalho muito semelhante ao trabalho feito antes. Como escrever código é demorado, é bom confiar no código publicado por outras pessoas sempre que possível. Em vez de copiar e colar, Python nos permite **importar** outro código, criando um **módulo** que contém todos os nomes criados por esse código. Esta é a forma de um programador atribuir o material ao criador original, como você sempre deve fazer!

Python inclui muitos módulos úteis que estão a apenas uma `importação` de distância. Veremos o módulo `math` como primeiro exemplo. O módulo `math` é extremamente útil no cálculo de expressões matemáticas em Python.

## 9.1. Importando constantes

Suponha que queiramos calcular com muita precisão a área de um círculo com raio de 5 metros. Para isso, precisamos da constante $\pi$, que é aproximadamente 3,14. Convenientemente, o módulo `math` tem `pi` definido para nós:

````python
import math
raio = 5
area_do_circulo = math.pi * raio**2
area_do_circulo
````

`pi` é definido dentro de `math`, e a forma como acessamos os nomes que estão dentro dos módulos é escrevendo o nome do módulo, depois um ponto e depois o nome do que queremos:

> <nome do módulo>.\<nome>

Para usar um módulo, devemos primeiro escrever a instrução `import <nome do módulo>`. Essa instrução cria um objeto de módulo com coisas como `pi` e então atribui o nome `math` a esse módulo. Acima fizemos isso para `math`.

**Exercício 2.10.** `math` também fornece o nome `e` para a base do logaritmo natural, que é aproximadamente 2,71. Calcule $e^{\pi}-\pi$, dando-lhe o nome `quase_vinte`.

````python
import math
quase_vinte = math.exp(math.pi) - math.pi
quase_vinte
````

![XKCD](http://imgs.xkcd.com/comics/e_to_the_pi_minus_pi.png)

**Módulos** podem fornecer outros itens nomeados além de constantes, incluindo **funções**. Por exemplo, `math` fornece o nome `sin` para a função seno. Já tendo importado `math`, podemos escrever `math.sin(3)` para calcular o seno de 3.

Observe que esta função senoidal considera seu argumento em [radianos](https://en.wikipedia.org/wiki/Radian), não em graus. 180 graus equivalem a $\pi$ radianos.

Um ângulo $\frac{\pi}{4}$-radianos (45 graus) forma um triângulo retângulo com base e altura iguais, como ilustrado abaixo. Se a hipotenusa (o raio do círculo na imagem) for 1, então a altura é $\sin(\frac{\pi}{4})$. Calcule essa altura usando `sin` e `pi` do módulo `math`. Dê ao resultado o nome `seno_de_pi_sobre_quatro`.

<img src="http://mathworld.wolfram.com/images/eps-gif/TrigonometryAnglesPi4_1000.gif">

Fonte: [Wolfram MathWorld](http://mathworld.wolfram.com/images/eps-gif/TrigonometryAnglesPi4_1000.gif)

````python
import math

# Calcular o seno de π/4
seno_de_pi_sobre_quatro = math.sin(math.pi / 4)

# Exibir o resultado
print(seno_de_pi_sobre_quatro)
````

Para sua referência, aqui estão mais alguns exemplos de funções do módulo `math`.

Observe como diferentes métodos aceitam diferentes números de argumentos. Frequentemente, a documentação do módulo fornecerá informações sobre quantos argumentos são necessários para cada método.

````python
# Calculando fatoriais
math.factorial(5)
````

````python
# Calculando logarítmos (logarítmo de 8 na base 2).
# O resultado é 3, já que 2^3 = 8.
math.log(8, 2)
````

````python
# Calculando raiz quadrada.
math.sqrt(5)
````

````python
# Calculando cossenos.
math.cos(math.pi)
````

### 9.1. A biblioteca 'math'


<a id = 'section6.1'></a>

Já sabemos fazer matemática básica em Python! Mas, e se precisarmos fazer contas matemáticas mais complexas, como potenciação, radiciação, valores absolutos, senos e cossenos, etc?<br>

Por sorte, Python tem várias bibliotecas para ajudar nós programadores. Uma biblioteca, de forma simples, é um conjunto de funções (veremos mais sobre isso mais tarde!).

Então, no caso de operações matemáticas mais complexas, podemos usar a biblioteca "math". Basta digitar `import math` no seu código, no começo do arquivo de código `.py` ou, se você estiver rodando o código em tempo de execução, digite antes de usar algo da biblioteca math pela primeira vez.<br><br>






Ao utilizar a biblioteca math, devemos nos atentar a alguns detalhes, como a sintaxe das funções matemáticas. Para utilizar uma função da biblioteca, escrevemos: `nome_da_funcao(valores)`.

A seguir, alguns exemplos de funções da math:
1. `ceil(valor)` - arredonda o número passado para cima;
2. `floor(valor)` - arredonda o número passado para baixo;
3. `factorial(valor)` - retorna o fatorial do número;
4. `pow(valor1, valor2)` - retorna o valor1 elevado ao valor2;
5. `sqrt(valor)` - retorna a raiz quadrada do valor;
6. `sin(valor), cos(valor), tan(valor)` - retorna, respectivamente, o seno, o cosseno e a tangente de um valor (em radianos).

Além disso, também temos as variáveis "pi" e "e", que vêm com a biblioteca. Por serem variáveis, você pode alterar seus valores do jeito que quiser, mas não é recomendado, uma vez que você pode voltar a precisar desses valores no futuro. Mas caso você tenha alterado essas variáveis e precise dos valores de novo, basta escrever:<br>
`pi = math.pi`<br>
`e = math.e`<br>

Caso você esteja curioso para ver o resto da documentação da biblioteca math, clique <a href = "https://docs.python.org/pt-br/3/library/math.html">aqui</a>.

Abaixo temos alguns exemplos usando funções da biblioteca math:

````python
import math
````

````python
# Isso é uma potenciação
print(pow(5, 3))
#se você preferir, pode fazer potenciações sem importar a math desse jeito:
print(5**3)
````

````python
# Isso é uma radiciação
print(math.sqrt(121))
#se você preferir, pode fazer radiciações sem importar a math desse jeito:
print(121**0.5)
````

````python
# Executa o fatorial de um número
print(math.factorial(5))
````

````python
# Retorna o cosseno de um número (em rad)
print(math.cos(0))
````

````python
# Arredonda o número para cima
print(math.ceil(4.25))
````

# 12. Desenvolvendo um jogo de Forca

Durante o curso, iremos desenvolver um jogo da forca. Ao final de cada aula, vamos aplicar o conteúdo ensinado para dar forma ao nosso jogo. O objetivo final é fazer com que o usuário seja capaz de adivinhar uma palavra secreta.

Ao longo desta aula, nós aprendemos a capturar uma entrada e exibí-la e já aprendemos alguns conceitos de tipos. Ambos os conteúdos são úteis para o esboço inicial de nosso jogo.

Agora, nós temos como objetivo capturar a palavra digitada de qualquer forma (usando caracteres maiúsculos ou minúsculos) e transformá-la inteiramente em uma palavra apenas com letras minúsculas.

Exemplo:

Se a entrada for ***InTROcOMp*** teremos como saída a palavra ***introcomp***.

### 12.1. Recebendo da entrada

Para recebermos a nossa palavra da entrada, devemos usar o "comando" `input`. Um bom nome para a variável que armazenará nossa palavra é `palavra`, uma vez que ele é significativo em nosso contexto. No fim, teremos:

````python
palavra = input("Qual a palavra a ser decifrada? ")
````

### 12.2. Tratando a _string_

Para que possamos utilizar melhor nossa _string_ no futuro, devemos tratá-la, de modo que ela não tenha espaços em branco nas extremidades e que seus caracteres sigam um padrão (no nosso caso, usaremos apenas caracteres minúsculos).

````python
palavra = input("Qual a palavra a ser decifrada? ") # O usuário irá digitar a entrada que ele quer dar para o programa

# Tratando a palavra:
# A função strip irá retirar os espaços do começo e do fim da nossa entrada digitada
# A função lower irá transformar todos os caracteres que estão em maiúsculo para minúsculo ( A -> a )
palavra = palavra.strip().lower()
````

### 12.3. Exibindo nossa _string_

Agora, utilizaremos o "comando" `print` para exibir nossa _string_ na saída padrão. Utilizaremos uma _f-string_.

**Observação**: Caso tenha dúvida no funcionamento desse recurso, veja a seção 7.4.

````python
palavra = input("Qual a palavra a ser decifrada? ") # O usuário irá digitar a entrada que ele quer dar para o programa

# Tratando a palavra:
# A função strip irá retirar os espaços do começo e do fim da nossa entrada digitada
# A função lower irá transformar todos os caracteres que estão em maiúsculo para minúsculo( A -> a )
palavra = palavra.strip().lower()



# Exibindo a palavra:
print(f'A palavra é "{palavra}".')
````

# 13. O que pode dar errado?


Erros acontecem com todo programador. Os programas apresentam diferentes tipos de erros, alguns estão relacionados a erros de sintaxe (geralmente acusados pelo interpretador) e outros a erros de semântica. Erros de semântica (ou seja, de lógica) são mais complicados pois dependem totalmente do programador para os encontrar e solucionar. Mas sempre tem uma forma de resolvê-los. Não desanimem e continuem programando!

Aqui estão alguns exemplos de erros comuns cometidos por programadores:

- Escrever algo da linguagem Python errado. Por exemplo, copie o trecho de código a seguir na celula abaixo e veja o que acontece: `pritn("Por que esse print não está funcionando? :c")`;
- Confundir letras maiúsculas e minúsculas nos nomes de variáveis: somaDosNumeros, SomaDosNumeros, somadosnumeros são 3 variáveis diferentes;
- Indentação e espaços em branco. Siga um modelo de tabulação consistente e único no seu script. Se você começou usando dois espaços, ou tabulação de 4 espaços ou seja o que for, mantenha-o. O interpretador, geralmente, não saberá o que fazer se você misturar vários de tipos de tabulação e espaçamento.



````python
print('Por que esse print não está funcionando? :c')
````

Existem estratégias para corrigir o código, as quais são especialmente úteis quando existem erros de semântica (o interpretador normalmente nos alerta sobre erros de sintaxe). Alguns exemplos consistem no uso de depurador/debug (ferramenta de rastreamento de erros), bem como na inserção da função `print()` para identificar o que o seu código está realizando em cada linha. Tais estratégias são utilizadas para que encontremos as falhas rapidamente. Saber onde elas estão quase sempre é a parte mais complicada.
