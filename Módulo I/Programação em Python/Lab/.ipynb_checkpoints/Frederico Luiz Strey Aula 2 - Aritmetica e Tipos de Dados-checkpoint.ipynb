# 2. O que  um programa?



Um programa  uma sequncia de instrues que especificam como o computador deve realizar uma tarefa. Essa tarefa pode ser algo como encontrar as razes de uma equao de segundo grau,  uma busca num texto para encontrar e substituir uma palavra, etc.  

Diferentemente de um algoritmo, um programa  escrito em alguma linguagem formal, chamada de **linguagem de programao**.
Da mesma forma como as pessoas se comunicam umas com as outras por meio de alguma linguagem (portugus, ingls, francs, etc.), precisamos de um jeito para nos comunicar com o computador. Para isso, utilizamos alguma linguagem de programao, como C, C++, Java, PHP, Lua, etc.

Neste curso, vamos aprender a linguagem de programao chamada **Python**.

Normalmente, as instrues que compem um programa variam de acordo com a linguagem de programao que voc est utilizando, mas  comum em praticamente todos os programas algumas instrues bsicas:

- Entrada: receber informaes, seja do teclado, de um arquivo, do mouse ou de um sensor de movimento numa cmera, por exemplo;
- Sada: enviar informaes, seja no terminal (caso no saiba o que  isso, fique tranquilo! Logo veremos mais sobre), em um arquivo, ou num display;
- Operaes Aritmticas: realizar operaes matemticas (soma, subtrao, diviso e etc), essencial para a maioria dos programas;
- Condicionais: checar condies e realizar instrues apropriadas;
- Repeties: realizar uma sequncia de comandos de forma repetitiva.

# 5. Variveis



Variveis so espaos na memria, destinados a armazenar informaes que sero utilizadas no futuro pelo computador. Pense como uma caixa que armazena algum determinado dado, e esta caixa recebe um nome (uma "etiqueta").

As variveis podem ter qualquer nome, respeitando algumas regras:

    1. O primeiro caracter precisa ser uma letra (A-Z, a-z) ou um underscore ( _ )
    2. Os demais caracteres podem ser letras, nmeros ou underscore

H algumas convenes sobre como devemos nomear variveis. Uma conveno possvel sugere que todo o nome da varivel deve ser escrito em letras minsculas e cada palavra deve ser separada por um *underscore* ("_"). Alguns exemplos esto presentes na tabela abaixo:

Nome modelo | Varivel seguindo essa conveno
------------|-------------------------
Dinheiro na conta|dinheiro_na_conta
Idade mnima|idade_minima
CPF do Cliente|cpf_cliente

**Deve-se evitar acentos e outros caracteres especiais ao criar variveis!**

**Dica:** tente atribuir nomes intuitivos e de fcil entendimento para as suas variveis! Isso facilita a construo do seu cdigo, bem como a sua futura reviso.

Alm disso, existem algumas palavras reservadas na linguagem, que no podem nomear variveis (ainda que respeitem as regras acima). Algumas dessas palavras so: True, False, continue, and, None, etc.

### 8.1. Funo `lower`



Transforma todos os caracteres que esto em maisculo para minsculo ( A -> a )

````python
frase = "ISSO  UMA STRING"

frase_modificada = frase.lower();

print(frase_modificada)
````

### 8.2. Funo `upper`



Transforma todos os caracteres que esto em minsculo para maisculo ( a -> A )

````python
frase = "est tudo pequeno"

frase_modificada = frase.upper();

print(frase_modificada)
````

### 8.3. Funo `strip`

Essa funo "ajeita" nossa string e  bem interessante para tratar entradas desse tipo de dado, uma vez que a chance de existirem caracteres especiais a mais diminui bastante.

Vejamos um exemplo sem `strip`.

````python
palavra = " IA-CD  "
palavra == "IA-CD"
````

Agora, veja como fica com o uso de `strip`:

````python
palavra = " IA-CD  ".strip()
palavra == "IA-CD"
````

Voc deve ter percebido que `strip` retirou os espaos em branco das extremidades da string.

### 8.4. _f-strings_

Alm de tratar nossas _strings_ com `upper`, `lower` e `strip`, existe uma maneira de utilizar uma _string_ como base para uma frase. Esse recurso foi adicionado na verso 3.6 de Python e  extremamente til.

Veja um exemplo:

````python
nome = input("Qual seu nome? ").strip()
saudacao = f"Bom dia, {nome}!"

print(saudacao)
````

Utilizamos a funo input() para podermos obter o que foi escrito pelo usurio. Perceba que o trecho onde existia um `{nome}` foi substitudo pelo contedo da varivel nome.  exatamente assim que utilizamos _f-strings_, basta inserir algo do tipo `{minha_variavel}` no meio de uma _string_. Uma outra caracterstica das _f-strings_  a necessidade da letra `f` antes do primeiro `"`.

### 8.5. Funo `format`

Por fim, vamos aprender como inserir variveis no meio de uma string (sequncia de letras e espaos) sem precisar separ-la com vrgulas ou operadores "+". Essa tcnica tambm  util para manter o cdigo mais legvel! Para colocarmos variveis no meio de uma string, basta colocar `{}` no exato local onde voc quer que a varivel aparea, e aps terminar a string, escreva `.format(nome_da_varivel)`. Um exemplo deve tornar as coisas mais simples:<br>

````python
imc = 22.54656
print("Uma pessoa com o IMC {} est dentro do seu peso ideal!".format(imc))
````

Achou esquisito? Geralmente, quando vamos apresentar nmeros com ponto flutuante como esse (chamados de **float** em Python), costumamos arredond-los para um nmero menor de casas decimais usando o especificador **.2f**, como no exemplo a seguir:

````python
imc = 22.54656
print("Uma pessoa com o IMC {:.2f} est dentro do seu peso ideal!".format(imc))
````

Podemos utilizar essa formatao tambm para imprimir strings:

````python
almoco = "strogonoff de frango"
````

````python
# bem fcil repetir o print que fizemos algumas clulas acima agora
print("quero {} no almoo!".format(almoco))
````

## 8.6. Mtodos de string

Strings podem ser transformadas usando **mtodos**, que so funes que envolvem uma string existente e alguns outros argumentos. Um exemplo  o mtodo `replace`, que substitui todas as instncias de alguma parte de uma string por alguma alternativa.

Um mtodo  invocado em uma string colocando um `.` aps o valor da string, depois o nome do mtodo e, finalmente, parnteses contendo os argumentos. Aqui est um esboo, onde os smbolos `<` e `>` no fazem parte da sintaxe; eles apenas marcam os limites das subexpresses.

> `<string>.<nome do mtodo>(<argumento>, <argumento>, ...)`

Tente prever o resultado desses exemplos e execute-os.

````python
# Substitundo uma letra
'Hello'.replace('o', 'a')
````

````python
# Substitundo uma sequncia de letras por outra (note que a sequncia se repete na palavra)
'hitchhiker'.replace('hi', 'ma')
````

Depois que uma varivel  vinculada a um valor de string, os mtodos tambm podem ser invocados nessa varivel (lembre-se, "varivel"  outra palavra para "nome"). A varivel `sharp` no muda neste caso, ento uma nova varivel  necessria para capturar o resultado.

````python
sharp = 'edged'
hot = sharp.replace('ed', 'ma')
print('sharp:', sharp)
print('hot:', hot)
````

````python
# Chamando duas vezes a funo replace
'train'.replace('t', 'ing').replace('in', 'de')
````

## 9.1. Importando constantes

Suponha que queiramos calcular com muita preciso a rea de um crculo com raio de 5 metros. Para isso, precisamos da constante $\pi$, que  aproximadamente 3,14. Convenientemente, o mdulo `math` tem `pi` definido para ns:

````python
import math
raio = 5
area_do_circulo = math.pi * raio**2
area_do_circulo
````

`pi`  definido dentro de `math`, e a forma como acessamos os nomes que esto dentro dos mdulos  escrevendo o nome do mdulo, depois um ponto e depois o nome do que queremos:

> <nome do mdulo>.\<nome>

Para usar um mdulo, devemos primeiro escrever a instruo `import <nome do mdulo>`. Essa instruo cria um objeto de mdulo com coisas como `pi` e ento atribui o nome `math` a esse mdulo. Acima fizemos isso para `math`.

**Exerccio 2.10.** `math` tambm fornece o nome `e` para a base do logaritmo natural, que  aproximadamente 2,71. Calcule $e^{\pi}-\pi$, dando-lhe o nome `quase_vinte`.

````python
import math
quase_vinte = math.exp(math.pi) - math.pi
quase_vinte
````

![XKCD](http://imgs.xkcd.com/comics/e_to_the_pi_minus_pi.png)

**Mdulos** podem fornecer outros itens nomeados alm de constantes, incluindo **funes**. Por exemplo, `math` fornece o nome `sin` para a funo seno. J tendo importado `math`, podemos escrever `math.sin(3)` para calcular o seno de 3.

Observe que esta funo senoidal considera seu argumento em [radianos](https://en.wikipedia.org/wiki/Radian), no em graus. 180 graus equivalem a $\pi$ radianos.

Um ngulo $\frac{\pi}{4}$-radianos (45 graus) forma um tringulo retngulo com base e altura iguais, como ilustrado abaixo. Se a hipotenusa (o raio do crculo na imagem) for 1, ento a altura  $\sin(\frac{\pi}{4})$. Calcule essa altura usando `sin` e `pi` do mdulo `math`. D ao resultado o nome `seno_de_pi_sobre_quatro`.

<img src="http://mathworld.wolfram.com/images/eps-gif/TrigonometryAnglesPi4_1000.gif">

Fonte: [Wolfram MathWorld](http://mathworld.wolfram.com/images/eps-gif/TrigonometryAnglesPi4_1000.gif)

````python
import math

# Calcular o seno de /4
seno_de_pi_sobre_quatro = math.sin(math.pi / 4)

# Exibir o resultado
print(seno_de_pi_sobre_quatro)
````

````python
# Calculando raiz quadrada.
math.sqrt(5)
````

````python
# Calculando cossenos.
math.cos(math.pi)
````

### 9.1. A biblioteca 'math'


<a id = 'section6.1'></a>

J sabemos fazer matemtica bsica em Python! Mas, e se precisarmos fazer contas matemticas mais complexas, como potenciao, radiciao, valores absolutos, senos e cossenos, etc?<br>

Por sorte, Python tem vrias bibliotecas para ajudar ns programadores. Uma biblioteca, de forma simples,  um conjunto de funes (veremos mais sobre isso mais tarde!).

Ento, no caso de operaes matemticas mais complexas, podemos usar a biblioteca "math". Basta digitar `import math` no seu cdigo, no comeo do arquivo de cdigo `.py` ou, se voc estiver rodando o cdigo em tempo de execuo, digite antes de usar algo da biblioteca math pela primeira vez.<br><br>






Ao utilizar a biblioteca math, devemos nos atentar a alguns detalhes, como a sintaxe das funes matemticas. Para utilizar uma funo da biblioteca, escrevemos: `nome_da_funcao(valores)`.

A seguir, alguns exemplos de funes da math:
1. `ceil(valor)` - arredonda o nmero passado para cima;
2. `floor(valor)` - arredonda o nmero passado para baixo;
3. `factorial(valor)` - retorna o fatorial do nmero;
4. `pow(valor1, valor2)` - retorna o valor1 elevado ao valor2;
5. `sqrt(valor)` - retorna a raiz quadrada do valor;
6. `sin(valor), cos(valor), tan(valor)` - retorna, respectivamente, o seno, o cosseno e a tangente de um valor (em radianos).

Alm disso, tambm temos as variveis "pi" e "e", que vm com a biblioteca. Por serem variveis, voc pode alterar seus valores do jeito que quiser, mas no  recomendado, uma vez que voc pode voltar a precisar desses valores no futuro. Mas caso voc tenha alterado essas variveis e precise dos valores de novo, basta escrever:<br>
`pi = math.pi`<br>
`e = math.e`<br>

Caso voc esteja curioso para ver o resto da documentao da biblioteca math, clique <a href = "https://docs.python.org/pt-br/3/library/math.html">aqui</a>.

Abaixo temos alguns exemplos usando funes da biblioteca math:

````python
import math
````

````python
# Isso  uma potenciao
print(pow(5, 3))
#se voc preferir, pode fazer potenciaes sem importar a math desse jeito:
print(5**3)
````

````python
# Isso  uma radiciao
print(math.sqrt(121))
#se voc preferir, pode fazer radiciaes sem importar a math desse jeito:
print(121**0.5)
````

````python
# Executa o fatorial de um nmero
print(math.factorial(5))
````

````python
# Retorna o cosseno de um nmero (em rad)
print(math.cos(0))
````

````python
# Arredonda o nmero para cima
print(math.ceil(4.25))
````

# 12. Desenvolvendo um jogo de Forca

Durante o curso, iremos desenvolver um jogo da forca. Ao final de cada aula, vamos aplicar o contedo ensinado para dar forma ao nosso jogo. O objetivo final  fazer com que o usurio seja capaz de adivinhar uma palavra secreta.

Ao longo desta aula, ns aprendemos a capturar uma entrada e exib-la e j aprendemos alguns conceitos de tipos. Ambos os contedos so teis para o esboo inicial de nosso jogo.

Agora, ns temos como objetivo capturar a palavra digitada de qualquer forma (usando caracteres maisculos ou minsculos) e transform-la inteiramente em uma palavra apenas com letras minsculas.

Exemplo:

Se a entrada for ***InTROcOMp*** teremos como sada a palavra ***introcomp***.

### 12.1. Recebendo da entrada

Para recebermos a nossa palavra da entrada, devemos usar o "comando" `input`. Um bom nome para a varivel que armazenar nossa palavra  `palavra`, uma vez que ele  significativo em nosso contexto. No fim, teremos:

````python
palavra = input("Qual a palavra a ser decifrada? ")
````

### 12.2. Tratando a _string_

Para que possamos utilizar melhor nossa _string_ no futuro, devemos trat-la, de modo que ela no tenha espaos em branco nas extremidades e que seus caracteres sigam um padro (no nosso caso, usaremos apenas caracteres minsculos).

````python
palavra = input("Qual a palavra a ser decifrada? ") # O usurio ir digitar a entrada que ele quer dar para o programa

# Tratando a palavra:
# A funo strip ir retirar os espaos do comeo e do fim da nossa entrada digitada
# A funo lower ir transformar todos os caracteres que esto em maisculo para minsculo ( A -> a )
palavra = palavra.strip().lower()
````

### 12.3. Exibindo nossa _string_

Agora, utilizaremos o "comando" `print` para exibir nossa _string_ na sada padro. Utilizaremos uma _f-string_.

**Observao**: Caso tenha dvida no funcionamento desse recurso, veja a seo 7.4.

````python
palavra = input("Qual a palavra a ser decifrada? ") # O usurio ir digitar a entrada que ele quer dar para o programa

# Tratando a palavra:
# A funo strip ir retirar os espaos do comeo e do fim da nossa entrada digitada
# A funo lower ir transformar todos os caracteres que esto em maisculo para minsculo( A -> a )
palavra = palavra.strip().lower()



# Exibindo a palavra:
print(f'A palavra  "{palavra}".')
````

# 13. O que pode dar errado?


Erros acontecem com todo programador. Os programas apresentam diferentes tipos de erros, alguns esto relacionados a erros de sintaxe (geralmente acusados pelo interpretador) e outros a erros de semntica. Erros de semntica (ou seja, de lgica) so mais complicados pois dependem totalmente do programador para os encontrar e solucionar. Mas sempre tem uma forma de resolv-los. No desanimem e continuem programando!

Aqui esto alguns exemplos de erros comuns cometidos por programadores:

- Escrever algo da linguagem Python errado. Por exemplo, copie o trecho de cdigo a seguir na celula abaixo e veja o que acontece: `pritn("Por que esse print no est funcionando? :c")`;
- Confundir letras maisculas e minsculas nos nomes de variveis: somaDosNumeros, SomaDosNumeros, somadosnumeros so 3 variveis diferentes;
- Indentao e espaos em branco. Siga um modelo de tabulao consistente e nico no seu script. Se voc comeou usando dois espaos, ou tabulao de 4 espaos ou seja o que for, mantenha-o. O interpretador, geralmente, no saber o que fazer se voc misturar vrios de tipos de tabulao e espaamento.



````python
print('Por que esse print no est funcionando? :c')
````

Existem estratgias para corrigir o cdigo, as quais so especialmente teis quando existem erros de semntica (o interpretador normalmente nos alerta sobre erros de sintaxe). Alguns exemplos consistem no uso de depurador/debug (ferramenta de rastreamento de erros), bem como na insero da funo `print()` para identificar o que o seu cdigo est realizando em cada linha. Tais estratgias so utilizadas para que encontremos as falhas rapidamente. Saber onde elas esto quase sempre  a parte mais complicada.
