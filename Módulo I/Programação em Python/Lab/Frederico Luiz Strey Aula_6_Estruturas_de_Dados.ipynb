## 3. Iterao e for-loops em listas

Como vimos na Aula 05, uma nica execuo de um conjunto de comandos dentro de um loop  chamada de **iterao**.

Existe uma propriedade muito interessante nas listas: elas so iterveis. Mas o que isso significa? Bem, quando falamos de estruturas iterveis, dizemos que elas so "percorrveis", ou seja, podemos acessar elemento a elemento de maneira ordenada, dentro de um loop (revise o contedo sobre loops na Aula 5 se achar necessrio).

### 3.1.  `for` em listas

Para que possamos iterar, utilizaremos a palavra-chave `for` em conjunto com uma lista e o operador `in`. Veja o exemplo a seguir:

````python
#lista declarada na seao 2.4!
for i in numeros:
    print(f"Nmero: {i}")
````

Veja como podemos utilizar cada elemento da lista, um de cada vez, de maneira simples. Para iterarmos uma lista, utilizamos a seguinte estrutura:

```python
for <varivel> in <lista>:
    <comandos>
```

A varivel recebe um elemento da lista de cada vez. Veja, na situao do exemplo anterior, na primeira iterao temos:
```python
for nome in nomes_dos_amigos:
    print(f"{nome}  meu amigo!")
```
Como essa  a primeira iterao, `nome` assumir o valor do primeiro elemento da lista `nomes_dos_amigos`. Dessa maneira, o comando ter o mesmo efeito que:
```python
print(f"{'Joo'}  meu amigo!")
```

Aps o fim do comando, teremos uma outra iterao (algo semelhante a rodadas de comandos, onde aps cada rodada o valor da varivel  atualizado). Nessa nova "rodada" teremos:
```python
print(f"{'Pedro'}  meu amigo!")
```
E por a vai... At que chegamos em `"Carlos"` e finalizamos todas as iteraes. Vejamos um cenrio prtico do uso de `for`:

````python
# Esse  um sistema de clculo de notas de alunos de uma turma:
alunos = ["Joo", "Pedro", "Henrique", "Manoel", "Carlos", "Gabriel"] # Lista de alunos
alunos_aprovados = 0

for aluno in alunos:
    nota = float(input(f"Qual a nota de {aluno}? "))

    # Notas iguais ou maiores que 7 significam aprovao
    if nota >= 7:
        print(f"{aluno} passou!")
        alunos_aprovados += 1
    else:
        print(f"{aluno} est de final!")

if alunos_aprovados == 1:
    passou = "passou"
else:
    passou = "passaram"
print(f"De {len(alunos)}, {alunos_aprovados} {passou} direto.")
````

Repare que para cada elemento da lista, o programa ir capturar a nota de derminado aluno e imprimir se ele passou ou no, ou seja, isso ocorrer 6 vezes.

## 5. Desenvolvendo nosso jogo

Agora que j entendemos um pouco mais sobre listas e loops, pode incrementar o jogo com a ajuda desses recursos. O estado do seu jogo ao fim da aula anterior deve ser algo parecido com:

````python
dificuldade = int(input("Escolha a dificuldade (de 1 a 10): "))

# Verificando se a dificuldade est correta:
if dificuldade > 10 or dificuldade < 1:
    quit()

# Definindo o nmero de tentativas com base na dificuldade:
tentativas_possiveis = 11 - dificuldade

# Perguntando a palavra para o usurio:
palavra = input("Qual a palavra a ser decifrada? ")

# Tratando a palavra:
palavra = palavra.strip().lower()

# Iniciando o jogo:
print("Pode comear!")

# Iniciando o loop:
finalizado = False
while not finalizado and tentativas_possiveis:
    chute = input("Chute uma palavra: ")

    # Verificando se o usurio acertou o chute:
    acertou = palavra == chute
    if acertou:
        print("Voc acertou!")
        finalizado = True
    else:
        print("Chute incorreto!")
        # Atualizando quantos chutes faltam
        tentativas_possiveis -= 1  # Lembre que isso e o mesmo que tentativas_possiveis = tentativas_possiveis - 1
        print(f"Voc pode errar mais {tentativas_possiveis} vezes!")

# finalizado significa que o problema foi resolvido:
if finalizado:
    print(f'A palavra era "{palavra}".')
````

````python
dificuldade = int(input("Escolha a dificuldade (de 1 a 10): "))

# Verificando se a dificuldade est correta:
if dificuldade > 10 or dificuldade < 1:
    quit()

# Definindo o nmero de tentativas com base na dificuldade:
tentativas_possiveis = 11 - dificuldade

# Perguntando a palavra para o usurio:
palavra = input("Qual a palavra a ser decifrada? ")

# Tratando a palavra:
palavra = palavra.strip().lower()

# Iniciando o jogo:
if palavra.isalpha(): # Verificando se a palavra no  proibida:
    print("Pode comear!")
    finalizado = False
else:
    print("Palavra proibida!")
    finalizado = True # Forma de forar parada do jogo.

# Iniciando o loop:
while not finalizado and tentativas_possiveis:
    chute = input("Chute uma palavra: ")

    # Verificando se o usurio acertou o chute:
    acertou = palavra == chute
    if acertou:
        print("Voc acertou!")
        finalizado = True
    else:
        print("Chute incorreto!")
        # Atualizando quantos chutes faltam
        tentativas_possiveis -= 1  # Lembre que isso e o mesmo que tentativas_possiveis = tentativas_possives - 1
        print(f"Voc pode errar mais {tentativas_possiveis} vezes!")

# finalizado significa que o problema foi resolvido:
if finalizado and palavra.isalpha(): # Se for parada por conta de erro, no exibir esse texto.
    print(f'A palavra era "{palavra}".')
````

# 7. Tuplas

As tuplas (**tuples**) so similares as listas, **porm uma vez criadas no podem ser modificadas**. So teis para a criao de valores que no podem ser mudados. Assim, o comando de atribuio de valor dado por '=' no pode ser utilizado para um elemento especfico da tupla aps a criao da mesma. Sua criao  ligeiramente diferente das listas e utiliza "()" ao invs de "\[\]". A indexao de valores tambm comea no valor zero e o acesso aos valores tambm utiliza o operador "[index]".

````python
tup = (3, 2, 5, 1)
cor = ("blue","red","yellow","green")
print(tup[1])
print(tup[1:])
print(tup[:3])
print(cor[-4:])
````

# 8. Conjuntos

Os conjuntos (**sets**)so declarados utilizando-se os operadores **'{}'**. Seus valores podem ser alterados como nas listas, porm estes no possuem uma indexao tal que a nica forma de acessar seus elementos  atravs dos nomes dos elementos. Para realizar a adio de um elemento usa-se **add** e mltiplos usa-se **update**. Para a remoo de um elemento usa-se **remove** ou **discard**. Para armazenar a unio de dois conjuntos em um terceiro e novo conjunto usa-se **union** e para armazenar o resultado da unio em um dos conjuntos usa-se **update**. O comando **issubset** verifica se um conjunto  subconjunto de outro conjunto e o resultado a ser retornado ser **True** se o conjunto for subconjunto de outro ou **false**, caso contrrio. O comando **intersection** retorna um subconjunto que  a interseco entre dois conjuntos.

````python
set1 = {"banana","ma","uva","laranja"}
set2 = {"green", "blue", "yellow", "red"}
print("banana" in set1) # Retorna verdadeiro (True) se houver o elemento banana em set1.
set1.add("mamo")
print(set1)
set2.update(["orange", "white", "black"])
print(set2)
set2.remove("orange")
print(set2)
set2.discard("white")
print(set2)
set3 = set1.union(set2)
print(set3)
set2.update(set1)
print(set2)
res = set1.issubset(set3)
print(res)
set4 = set3.intersection(set1)
print(set4)
````

# 9. Dicionrios

Os dicionrios (**dictionaries**)  uma coleo sem ordenao, mutvel e com indexao. Seus elementos so organizados em um estrutura de par (chave: valor) e so separados por vrgulas. A chave pode ser qualquer tipo bsico, e o valor pode ser um tipo bsico ou uma coleo. A declarao de um dicionrio utiliza o operador '{}'. Para acessar um valor ('value')  necessrio fornecer o nome da chave correspondente ('key') utilizando o operador '[]'. Para se saber quais as chaves de um dicionrio d usa-se **d.keys()**, os valores usa-se **d.values()**, e chaves e valores usa-se **items**. Para se modificar o valor val com val2 na posio **pos** de uma lista **l1** associada a uma chave **'k'** de um
dicionrio **d** usa-se **d['k'][pos] = val2**.

````python
cores = {1: 'blue', 2: 'red', 3: 'yellow', 4: 'green'}
dados = {'x': [1, 2, 3, 4, 5], 'y': [1, 4, 9, 16, 25]}

print(cores[1])
print(dados['x'])
print(cores.keys())
print(dados.values())
print(cores.items())
print(cores)
cores["4"] = 'black'
print(cores)
print(dados)
dados["y"][0] = 0
print(dados)
````
