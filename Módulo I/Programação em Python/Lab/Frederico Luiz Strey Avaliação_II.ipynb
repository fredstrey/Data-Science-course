## Exerccio 9.1.

## Exerccio 9.2.
Faa um programa que pea o nome e a altura de 5 pessoas, armazene cada informao no seu respectivo vetor, e ento imprima o nome e a altura na ordem inversa  ordem lida.

**Entrada:**\
joana - 1.55\
cleber - 1.65\
rodrigo - 1.78\
jorge - 1.87\
carlos - 1.90\
**Sada:**\
Nome: carlos - Altura: 1.90\
Nome: jorge - Altura: 1.78\
Nome: rodrigo - Altura: 1.78\
Nome: cleber - Altura: 1.65\
Nome: joana - Altura: 1.55

`````python
def altura_5pessoas(n):
  lista=[]
  for x in range(n):
    nome_altura=input('digite: "nome - altura"')
    nome, altura = nome_altura.split(' - ')
    lista.append([nome.strip(), float(altura.strip())])#cria uma lista com [nome,altura] e junta numa lista com append()
  return lista

def imprime_inverso(n):
    lista = altura_5pessoas(n)
    for i in range(n - 1, -1, -1):
      nome, altura = lista[i]
      print(f"Nome: {nome} - Altura: {altura}")
imprime_inverso(5)
`````

## Exerccio 9.3.
Faa um programa que converta da notao de 24 horas para a notao de 12 horas. A entrada  dada em dois inteiros. Deve haver uma funo para fazer a converso e para a sada no formato $H:M$. Essa funo ter que escrever $PM$ ou $AM$, a depender do horrio de entrada. Inclua um loop que permita que o usurio repita esse clculo para novos valores de entrada todas as vezes que desejar.

**Entrada:**\
9\
35\
**Sada:**\
09:35 AM\
\
**Entrada:**\
14\
56\
**Sada:**\
02:56 PM

`````python
def converssor():
    hora = int(input('digite as horas: '))
    minuto = int(input('digite os minutos: '))

    if hora < 12:
        if hora == 0:
            hora = 12
        print(f'{hora}:{minuto} AM')
    elif hora > 12:
        hora -= 12
        print(f'{hora}:{minuto:0} PM')
    elif hora == 12:
        print(f'{hora}:{minuto:0} PM')

converssor()
`````

## Exerccio 9.4.
Construa uma funo que receba uma string como parmetro e devolva outra string com os carateres embaralhados de forma aleatoria (utilize a biblioteca random!). Padronize em sua funo que todos os caracteres sero devolvidos em caixa alta ou caixa baixa, independentemente de como foram digitados.

Um exemplo de uso da biblioteca random que pode auxiliar:
````

# Gera um nmero aleatrio entre 1 e 10

````
**Entrada:**\
Python\
**Sada:**\
yphtno\
\
**Entrada:**\
Introcomp\
**Sada:**\
tcpminoor

`````python
import random

def embaralhar_string(string):

    string = string.lower()
    lista_caracteres = list(string)
    random.shuffle(lista_caracteres)#embaralha os elementos da lista
    nova_string = ''.join(lista_caracteres)#junta os elementos da lista pra formar a nova string, '' indica que os elementos sero concatenados sem espao entre eles

    return nova_string

print(embaralhar_string("Python"))
print(embaralhar_string("Introcomp"))
`````
