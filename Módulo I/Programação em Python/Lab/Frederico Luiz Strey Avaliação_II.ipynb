## Exercício 9.1.
Faça um programa que receba uma frase e conte o número de letras e algarismos.

Algarismos:1\

helloworld 123\
**Saída:**\

Algarismos:3

## Exercício 9.2.
Faça um programa que peça o nome e a altura de 5 pessoas, armazene cada informação no seu respectivo vetor, e então imprima o nome e a altura na ordem inversa à ordem lida.

**Entrada:**\
joana - 1.55\
cleber - 1.65\
rodrigo - 1.78\
jorge - 1.87\
carlos - 1.90\
**Saída:**\
Nome: carlos - Altura: 1.90\
Nome: jorge - Altura: 1.78\
Nome: rodrigo - Altura: 1.78\
Nome: cleber - Altura: 1.65\
Nome: joana - Altura: 1.55

`````python
def altura_5pessoas(n):
  lista=[]
  for x in range(n):
    nome_altura=input('digite: "nome - altura"')
    nome, altura = nome_altura.split(' - ')
    lista.append([nome.strip(), float(altura.strip())])#cria uma lista com [nome,altura] e junta numa lista com append()
  return lista

def imprime_inverso(n):
    lista = altura_5pessoas(n)
    for i in range(n - 1, -1, -1):
      nome, altura = lista[i]
      print(f"Nome: {nome} - Altura: {altura}")
imprime_inverso(5)
`````

## Exercício 9.3.
Faça um programa que converta da notação de 24 horas para a notação de 12 horas. A entrada é dada em dois inteiros. Deve haver uma função para fazer a conversão e para a saída no formato $H:M$. Essa função terá que escrever $PM$ ou $AM$, a depender do horário de entrada. Inclua um loop que permita que o usuário repita esse cálculo para novos valores de entrada todas as vezes que desejar.

**Entrada:**\
9\
35\
**Saída:**\
09:35 AM\
\
**Entrada:**\
14\
56\
**Saída:**\
02:56 PM

`````python
def converssor():
    hora = int(input('digite as horas: '))
    minuto = int(input('digite os minutos: '))

    if hora < 12:
        if hora == 0:
            hora = 12
        print(f'{hora}:{minuto} AM')
    elif hora > 12:
        hora -= 12
        print(f'{hora}:{minuto:0} PM')
    elif hora == 12:
        print(f'{hora}:{minuto:0} PM')

converssor()
`````

## Exercício 9.4.
Construa uma função que receba uma string como parâmetro e devolva outra string com os carateres embaralhados de forma aleatoria (utilize a biblioteca random!). Padronize em sua função que todos os caracteres serão devolvidos em caixa alta ou caixa baixa, independentemente de como foram digitados.

Um exemplo de uso da biblioteca random que pode auxiliar:
````

# Gera um número aleatório entre 1 e 10
numero_aleatorio = random.randint(1, 10)
````
**Entrada:**\
Python\
**Saída:**\
yphtno\
\
**Entrada:**\
Introcomp\
**Saída:**\
tcpminoor

`````python
import random

def embaralhar_string(string):

    string = string.lower()
    lista_caracteres = list(string)
    random.shuffle(lista_caracteres)#embaralha os elementos da lista
    nova_string = ''.join(lista_caracteres)#junta os elementos da lista pra formar a nova string, '' indica que os elementos serão concatenados sem espaço entre eles

    return nova_string

print(embaralhar_string("Python"))
print(embaralhar_string("Introcomp"))
`````
