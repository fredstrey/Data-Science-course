### 1. Paradigmas de Programação

Antes de falarmos sobre o paradigma de programação orientado a objeto, é preciso entender *o que são* paradigmas de programação.

Os paradigmas de programação podem ser vistos como um meio, metodologia ou maneira de se classificar o uso de uma linguagem a partir de suas **funcionalidades**, ou, mais especificamente, das estruturas que ela dispõe para resolução de problemas.

Entre os paradigmas existentes, podemos citar o procedural, o funcional, o estruturado, o imperativo e o orientado a objetos. Conhecer os vários paradigmas de programação permite abordagens mais objetivas e otimizadas na confecção de nossos códigos.

No mundo do desenvolvimento de softwares, existem linguagens conhecidas como multiparadigmas; como o nome sugere, essas linguagens suportam mais de um paradigma, como é o caso de Python. Um dos paradigmas suportados por ele é o de POO (Programação Orientada a Objetos).

É importante notar que, no decorrer do curso até agora, utilizamos apenas o paradigma imperativo/procedural.

#### 1.1. Programação Imperativa

O paradigma de programação imperativa é fortemente baseada no modelo apresentado pelo matemático *Alan Turing*, denominado "A máquina de Turing", no qual o programador especifica passos que devem ser seguidos em ordem pelo computador.

Quase sempre, em conjunto com a programação imperativa, utilizamos o **paradigma de programação procedural**, na qual o programador descreve as ações, passo-a-passo do que deve ser executado, isto é, ele descreve procedimentos.

Esses são os únicos paradigmas que utilizamos durante o curso até agora.

#### 1.2. Programação Orientada a Objetos

Originalmente sugerida por *Alan Kay* por volta de 1969 e desenvolvida por outros grandes pesquisadores ao longo do tempo, a **programação orientada a objetos** utiliza conceitos que visam facilitar a legibilidade, manutenção e reaproveitamento de _softwares_.

O intuito da sua criação também foi o de aproximar o manuseio das estruturas de um programa ao manuseio das coisas do mundo real, daí o nome "objeto" como algo genérico, que pode representar qualquer coisa tangível (pessoas, objetos, automóveis, empresas, funcionários, animais, ferramentas, etc). Sendo assim, programação orientada a objetos tem como foco principal o item com o qual estamos lidando.

Considerando o exemplo anterior do carro, usando o paradigma de orientação a objetos, teríamos um objeto `carro` e tal objeto teria suas **características específicas**, como por exemplo sua velocidade. Usando a expressão `carro.velocidade = 60` podemos atribuir o valor 60 à velocidade do carro. Podemos também definir uma classe de objetos denominada `Veiculos` que exibe propriedades comuns a todos os veículos, como por exemplo, velocidade, direção, aceleraçào, etc. Como mostraremos, isso facilita a criação de carros, caminhões e motocicletas.

Alguns recursos da orientação a objetos no curso podem parecer familiares, considerando que já foram mostradas ao longo do curso: um exemplo é a "função" `lista.append()`; veremos que a notação de ponto '`.`' é utilizada para acessar propriedades/atributos de um certo objeto.

### 2.1. Abstração

Lembre-se que queremos trabalhar com conceitos reais de objetos, então precisamos estudar e entender o que tal representação de um objeto real realizará dentro do nosso sistema.

Abstração é o ponto de partida para se trabalhar com orientação a objeto. Imagine que você vai escrever um texto para um blog de automóveis e precisa colocar todas as características de um carro nesse texto. Você provavelmente falaria sobre a cor, o modelo, a marca, o ano de fabricação, número de portas, potência, etc. Isso é um bom exemplo de abstração, pois nele nos concentramos nas propriedades e ações que são relevantes para nosso problema.

Se estivéssemos implementando um sistema para uma fábrica de carros, teríamos outras propriedades e ações de interesse (material, peso, custo, etc). Isso é a abstração, você deixa de trabalhar com todas as variáveis e se preocupa apenas com as **características relevantes** do seu **objeto**.

Em orientação a objetos, as *características* do objeto são chamadas **propriedades**, e suas *ações* são chamadas **métodos**. Os métodos seriam as "funções" das classes. Portanto temos, como exemplos, a propriedade `carro.placa` e o método `carro.acelerar()`.

### 2.1.1. Construtores

Ao criar uma classe `Carro`, por exemplo, com suas propriedades estabelecidas, é preciso definir o método construtor daquela classe.

O construtor é o método que, como o próprio nome diz, constrói o objeto de uma determinada classe. Ele funciona **parecido** com uma função e geralmente é responsável pela alocação de recursos necessários ao funcionamento do objeto, além da definição inicial das variáveis de estado (atributos/propriedades). É semelhante a quando criamos uma variável do tipo int ou string, mas com o diferencial de que agora essa variável terá várias características e "funções" próprias pertencentes àquele tipo.

O construtor faz a inicialização do objeto instanciado.

```
#Pseudocódigo
classe Carro:
  Propriedades:
    marca,
    modelo,
    placa,
    velocidade,
    velocidade_max
  Métodos:
    construtor()

```

### 2.2. Herança

Ainda em nosso exemplo da classe `Carro`, podemos definir um automóvel como herdeiro da classe `Veículos`. Mas como assim herdeira? Veja bem, se pensarmos nos vários tipos de véiculos (carros, caminhões, motos, vans, etc), todos eles possuem características em comum. Entre elas podemos citar:

* Placa
* Ano
* Combustível
* Cor
* Proprietário

Portanto, podemos definir uma classe mais genérica chamada `Veículo` da seguinte forma:

```
#Pseudocódigo
    classe Veículo:
        Propriedades:
            Ano,
            Placa,
            Proprietário
        Métodos:
            acelerar(),
            ligar_farol(),
            frear(),
            buzinar()
```

Podemos agora recriar a classe `Carro` como **herdeira** da classe `Veículo`:

```
#Pseudocódigo
    classe Carro (herdeira de Veículo):
        Métodos:
            abrir_porta(),
            abrir_janela()
```

Perceba que não declaramos nenhuma propriedade. Com a herança, a classe `Carro` **herda** todos os métodos e propriedades da classe `Veículo`. Também é possível adicionar métodos ou propriedades que não estejam definidas nas classe herdada, como é o caso de `carro.abrir_porta()` e `carro.abrir_janela()`.

Caso algum método tenha o mesmo nome, mas funcione de maneira diferente (como por exemplo uma bicicleta não acende o farol da mesma forma que um carro ou até mesmo uma moto), basta apenas **sobrescrever** os métodos que terão comportamento diferente, isto é, declarar o método novamente dentro da classe, com o mesmo nome da que precisa ser sobrescrita, e declarar as ações que servem aquele objeto.

### 3.1. Criando classes e seus objetos

Para criar uma classe em _Python_ basta utilizar a palavra-chave `class` seguida do nome da classe. No nosso caso seria algo como:

```python

```

Recomendamos que escreva, através de comentário, as propriedades da classe. Dessa maneira será mais fácil lembrar delas mais tarde.

Em seguida definimos o construtor com `__init__`.

```python
class Personagem:
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
    """
    def __init__(self, nome, ataque, vida):
        self.nome = nome
        self.ataque = ataque
        self.vida = vida
```


*   ``__init__`` é o nome reservado para o construtor de uma classe. Ele funciona parecido com uma função e seus parâmetros são respectivamente (*self*, *propriedade 1*, *propriedade 2*).

* A palavra "_self_" significa algo como "próprio" ou "auto". Ao trabalharmos com classes, usar esse termo como parâmetro de uma função indica que, quando aquela função for chamada, propriedades do próprio objeto serão acessadas (podendo ser modificadas). Com a prática vai tudo ficar bem claro!

* O `.` é usado para **acessar determinada característica/propriedade** de algum objeto especificado.

Dessa forma, quando colocamos `self.nome`, estamos acessando o nome do próprio objeto onde a função está declarada. Com isso, ao utilizar o init apresentado, sabemos que estamos modificando as propriedades `nome`, `ataque` e `vida` do nosso objeto, a fim de **inicializar** ele.

Vamos declarar nossa classe:

````python
#Execute essa célula para declarar a classe Personagem!
class Personagem:
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
    """
    def __init__(self, nome, ataque, vida):
        self.nome = nome
        self.ataque = ataque
        self.vida = vida
````

Vamos realizar alguns teste fazendo a instaciação, ou seja, a declaração *e* criação do objeto.

````python
goku = Personagem("Goku", 5500, 1000)
````

parênteses, estamos nos referindo ao construtor.

Além disso, vale notar que ao realizar a instanciação, não é necessário passar nada como argumento para o parâmetro `self`. Considerando que ele se refere ao próprio objeto, definir "qual é o objeto" seria redundante.

````python
#Exibindo nosso personagem:
print(f"nome: {goku.nome} ataque: {goku.ataque} vida: {goku.vida}")
````

### 3.2. Métodos

Como dito antes, métodos são as "funções" das classes. Vamos colocar a mão na massa e criar o método `apresenta_personagem`, para não precisamos de usar "prints" de maneira repetitiva:

Agora, ficou muito mais simples exibir informações sobre nossos personagens. Contudo, existe uma maneira melhor de implementar essa funcionalidade.

Estamos falando de um método chamado `__str__`. Ao exibir um objeto com `print`, o _Python_ invoca esse método, que acessa o objeto e as suas propriedades, retornando uma string no formato definido em `__str__`.

Além de `__init__` e `__str__`, outros métodos com nomes "especiais" podem melhorar sua classe; esses métodos podem ser encontrados [aqui](https://docs.python.org/pt-br/3/library/functions.html).

````python
class Personagem:
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
    """
    def __init__(self, nome, ataque, vida):
        self.nome = nome
        self.ataque = ataque
        self.vida = vida

    def __str__(self):
        string_apresentacao = f"Olá! Meu nome é {self.nome}. Meu poder de ataque é {self.ataque} e tenho {self.vida} de vida!"
        return string_apresentacao
````

### 3.3. Herança

Agora que já vimos como funciona a criação de uma classe e seus métodos, vejamos como isso pode ser aplicado com o conceito de herança.

Para isso, vamos imaginar uma classe mais específica de personagem: a classe `Defensor`. Antes disso, vamos declarar nossa classe-mãe:

````python
class Personagem:
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
    """
    def __init__(self, nome, ataque, vida):
        self.nome = nome
        self.ataque = ataque
        self.vida = vida

    def __str__(self):
        string_apresentacao = f"Olá! Meu nome é {self.nome}. Meu poder de ataque é {self.ataque} e tenho {self.vida} de vida!"
        return string_apresentacao

    def esta_vivo(self):
        return self.vida > 0
````

Agora, podemos criar classes mais específicas, que herdem os métodos e os atributos da classe `Personagem`. Além dos atributos de um personagem genérico, a classe `Defensor` terá um escudo, o que lhe confere vida extra!

````python
class Defensor(Personagem):
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
        escudo: quantidade extra de vida(int ou float)
    """
    def __init__(self, nome, ataque, vida, escudo):
        super().__init__(nome, ataque, vida) #esta chamada serve para invocar o método init da classe-mãe!
        self.escudo = escudo

    def __str__(self):
        return super().__str__() + f"Sou um defensor com {self.escudo} de escudo!" #Aqui, unimos à apresentação da classe-mãe informações sobre o escudo
````

````python
gragas = Defensor("Gragas", 300, 5000, 2000)

if gragas.esta_vivo(): #repare que podemos usar métodos da classe mãe sem declará-los dentro de "Defensor"!
    print(gragas)
else: #se a primeira linha for "gragas = Defensor("Gragas", 300, 0, 2000)", essa condicao sera satisfeita
    print(f"R.I.P {gragas.nome} :(")
````

Não precisamos reescrever todos os métodos de `Personagem` para o tipo `Defensor`, afinal ele _herdou_ todas as propriedades e métodos de sua classe-mãe. Contudo, vale ressaltar que ao usarmos algum método da classe-mãe na definição dos métodos da classe-fillha, devemos lançar mão do termo "super()." seguido do nome desse método - foi o que fizemos no init e no str. Por fim, repare que não há conflito ao usarmos funções de nomes iguais em classes que possuem relação de herança, pois é permitido *sobrescrever* funções.

**OBS:** Note que a classe herdeira, ou seja, a classe derivada, tem acesso a todas as propriedades e métodos da classe principal, mas **o contrário não é válido**. A classe-mãe (no nosso exemplo, `Personagem`, não tem acesso à propriedade `escudo` da classe `Defensor`).
Pense como se fosse uma herança mesmo: a pessoa que recebeu a herança tem acesso a todo o dinheiro da pessoa que morreu, mas a pessoa que morreu não tem acesso a nada daquele que recebeu :(

**Exercício 8.2.**

Agora que aprendemos sobre criação de métodos, nos ajude a criar o método `dobrar_vida`, que deve multiplicar a vida do nosso defensor por 2!

````python
class Defensor(Personagem):
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
        escudo: quantidade extra de vida(int ou float)
    """
    def __init__(self, nome, ataque, vida, escudo):
        super().__init__(nome, ataque, vida)
        self.escudo = escudo

    def __str__(self):
        return super().__str__() + f"Sou um defensor com {self.escudo} de escudo!"

    def dobrar_vida(self):
        self.vida *= 2
````

## 4. Colocando em Prática

Agora que temos uma boa base dos conceitos principais de orientação a objetos, podemos desenvolver um projeto mais robusto, um pouco mais próximo dos programas que faremos daqui em diante. Nosso objetivo nessa subseção, é usar as classes que fizemos e simular um round entre duas equipes, que serão constituídas de um atacante e um defensor. ### 4.1. Classe Personagem

Essa classe já está pronta! ela é organizada assim:

```
classe Personagem:
    Propriedades:
        nome,
        ataque,
        vida
    Métodos:
        construtor
        str
        esta_vivo
        recebe_dano
```

````python
class Personagem:
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
    """
    def __init__(self, nome, ataque, vida):
        self.nome = nome
        self.ataque = ataque
        self.vida = vida

    def __str__(self):
        string_apresentacao = f"Olá! Meu nome é {self.nome}. Meu poder de ataque é {self.ataque} e tenho {self.vida} de vida!"
        return string_apresentacao

    def recebe_dano(self, ataque):
        self.vida -= ataque
        if(self.vida < 0):
          self.vida = 0
    def esta_vivo(self):
        """
        Verifica se o personagem está vivo.
        """
        return self.vida > 0
````

A equipe contém um Atacante e um Defensor, que podemos enxergar como objetos de classes derivadas da classe Personagem. Já conhecemos a classe Defensor, mas precisanos que ela seja um pouco mais robusta:

````python
class Defensor(Personagem):
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
        escudo: quantidade extra de vida(int ou float)
    """
    def __init__(self, nome, ataque, vida, escudo):
        super().__init__(nome, ataque, vida)
        self.escudo = escudo

    def __str__(self):
        return super().__str__() + f"Sou um defensor com {self.escudo} de escudo!"

    '''
    Essa funcao processa um ataque recebido de outro personagem levando
    em conta o escudo do defensor; para isso, sobrescrevemos a função
    recebe_dano
    '''
    def recebe_dano(self, ataque):
        if self.escudo > 0:
          if (self.escudo - ataque) >= 0: #verifica se o escudo vai aguentar todo o dano recebido
              self.escudo -= ataque
          else: #caso em que o escudo aguentou apenas uma parte do dano
              self.vida -= (ataque - self.escudo)
              self.escudo = 0

              if(self.vida < 0):
                self.vida = 0
        else:
            self.vida -= ataque

            if(self.vida < 0):
                self.vida = 0
````

**Exercício 8.3.**

Agora, implemente a classe `Atacante`. Além dos atributos de um personagem genérico, ele tem uma taxa (que varia de 0 a 1) de crítico, que representa uma porcentagem (referente à sua própria força) que deve ser adicionada em seu ataque.

Siga a seguinte estrutura:
```
classe Atacante:
    Propriedades:
        nome,
        forca,
        vida,
        taxa_critico
    Métodos:
        construtor (init)
        impressão (str)
        calcula_ataque
```

Dica: Observe o exemplo da classe `Defensor` para realizar essa tarefa!

````python
class Atacante(Personagem):
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
        taxa_critico: porcentagem (0 a 1) de chance crítica (float)
    """
    def __init__(self, nome, ataque, vida, taxa_critico):
        super().__init__(nome, ataque, vida)
        self.taxa_critico = taxa_critico

    def __str__(self):
        return super().__str__() + f" Tenho {self.taxa_critico * 100}% de chance de crítico!"

    def calcula_ataque(self):
        """
        Calcula o ataque levando em consideração a taxa de crítico.
        """
        ataque_critico = self.ataque * (1 + self.taxa_critico)
        return ataque_critico

    def recebe_dano(self, dano):
        """
        Recebe dano e atualiza a vida do atacante.
        Se a vida cair abaixo de zero, ela é ajustada para zero.
        """
        self.vida -= dano
        if self.vida < 0:
            self.vida = 0
````

### 4.2. Classe Equipe

Nossa classe Equipe deverá ter um nome e dois personagens: um Atacante e um Defensor. Essa escolha facilita as coisas, mas poderíamos fazer uma equipe com vários personagens, sem distinção entre Atacante e Defensor: bastaria que criássemos uma coleção de personagens! (lembre-se que uma coleção pode ser uma lista, tupla ou dicionário).


Desse modo, a organização dessa classe deve ser mais ou menos assim:
```
classe Equipe:
    Propriedades:
        nome,
        atacante,
        defensor
    Métodos:
        construtor (init)
        impressão (str)
        foi_derrotada
        realiza_ataque
```

Obs.: A função `foi_derrotada` deve verificar se existe algum personagem da equipe vivo (e, se sim, retornar `False`); já a `realiza_ataque` deve usar os ataques dos membros da equipe contra um personagem.

**Exercício 8.4.**

Com base nas informações dadas até aqui, complete nossa classe `Equipe`:

````python
class Equipe:
    """
    Classe para representar uma equipe de personagens.

    Propriedades:
        nome: Nome da equipe (str)
        atacante: Personagem do tipo Atacante
        defensor: Personagem do tipo Defensor
    """

    def __init__(self, nome: str, atacante: Atacante, defensor: Defensor):
        self.nome = nome
        self.atacante = atacante
        self.defensor = defensor

    def __str__(self) -> str:
        return f"Equipe {self.nome}: {self.atacante.nome} (Atacante) e {self.defensor.nome} (Defensor)"

    def foi_derrotada(self) -> bool:
        """
        Verifica se a equipe foi derrotada, ou seja, se não há personagens vivos.
        """
        return self.atacante.vida <= 0 and self.defensor.vida <= 0

    def realiza_ataque(self, alvo: Personagem) -> None:
        """
        Realiza um ataque do atacante da equipe contra um alvo.
        """
        if self.atacante.esta_vivo():
            ataque = self.atacante.calcula_ataque()
            alvo.recebe_dano(ataque)
            print(f"{self.atacante.nome} atacou {alvo.nome} e causou {ataque:.2f} de dano!")
        else:
            print(f"{self.atacante.nome} não pode atacar porque está derrotado!")
````

**Exercício 8.5.**

Implemente o método `realiza_ataques`. Para isso, leve em conta que os ataques só devem ser direcionados ao atacante da outra equipe se o defensor estiver morto.

````python
class Equipe:
    """
    Classe para representar uma equipe de personagens.

    Propriedades:
        nome: Nome da equipe (str)
        atacante: Personagem do tipo Atacante
        defensor: Personagem do tipo Defensor
    """

    def __init__(self, nome: str, atacante: Atacante, defensor: Defensor):
        self.nome = nome
        self.atacante = atacante
        self.defensor = defensor

    def __str__(self) -> str:
        return f"{self.nome} - Atacante: {self.atacante.nome} (Vida: {self.atacante.vida}) | Defensor: {self.defensor.nome} (Vida: {self.defensor.vida}, Escudo: {self.defensor.escudo})"

    def foi_derrotada(self) -> bool:
        """
        Verifica se a equipe foi derrotada, ou seja, se não há personagens vivos.
        """
        return self.atacante.vida <= 0 and self.defensor.vida <= 0

    def realiza_ataque(self, alvo: Personagem) -> None:
        """
        Realiza um ataque do atacante da equipe contra um alvo.
        """
        if self.atacante.esta_vivo():
            ataque = self.atacante.calcula_ataque()
            alvo.recebe_dano(ataque)
            print(f"{self.atacante.nome} atacou {alvo.nome} e causou {ataque:.2f} de dano!")
        else:
            print(f"{self.atacante.nome} não pode atacar porque está derrotado!")

    def realiza_ataques(self, outra_equipe: 'Equipe') -> None:
        """
        Realiza ataques contra a outra equipe.
        Se o defensor da outra equipe estiver vivo, o ataque é direcionado a ele.
        Se o defensor estiver derrotado, o ataque é direcionado ao atacante da outra equipe.
        """
        if outra_equipe.defensor.esta_vivo():
            print(f"{self.nome} ataca o defensor {outra_equipe.defensor.nome} da equipe {outra_equipe.nome}!")
            self.realiza_ataque(outra_equipe.defensor)
        else:
            print(f"{self.nome} ataca o atacante {outra_equipe.atacante.nome} da equipe {outra_equipe.nome}!")
            self.realiza_ataque(outra_equipe.atacante)

````

À essa altura, já criamos um projeto bem legal. Vamos testar:

````python
if __name__ == "__main__":
    # Criação de personagens
    atacante1 = Atacante("Legolas", ataque=15, vida=100, taxa_critico=0.5)
    defensor1 = Defensor("Gimli", ataque=10, vida=120, escudo=50)

    atacante2 = Atacante("Nazgûl", ataque=15, vida=100, taxa_critico=0.5)
    defensor2 = Defensor("Orc", ataque=5, vida=70, escudo=20)

    # Criação das equipes
    equipe1 = Equipe("Sociedade do Anel", atacante1, defensor1)
    equipe2 = Equipe("Exército das Sombras", atacante2, defensor2)

    # Impressão das equipes
    print(equipe1)
    print(equipe2)

    # Realiza ataques entre as equipes
    while not equipe1.foi_derrotada() and not equipe2.foi_derrotada():
        equipe1.realiza_ataques(equipe2)
        if not equipe2.foi_derrotada():
            equipe2.realiza_ataques(equipe1)

    # Verificar qual equipe foi derrotada
    if equipe1.foi_derrotada():
        print(f"A equipe {equipe1.nome} foi derrotada!")
    if equipe2.foi_derrotada():
        print(f"A equipe {equipe2.nome} foi derrotada!")
````

## 5. O que pode dar errado?

Como pudemos perceber, nossos códigos ficaram um pouco mais complexos - com isso, surgem novos possíveis erros. O exemplo abaixo visa ilustrar isso:

A classe abaixo representa um triângulo retângulo. Como bem sabemos, um triângulo retângulo de catetos com, respectivamente, $a$ e $b$ de comprimento e hipotenusa de $c$ de comprimento, deve seguir a fórmula:
$c^2 = a^2 + b^2$

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """

    #adicionamos `__` antes dos nomes da variáveis para identificar que são privadas
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)

    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.__cateto_a, self.__cateto_b, self.__hipotenusa)

    def cateto_a_getter(self):
        return self.__cateto_a

    def cateto_b_getter(self):
        return self.__cateto_b

    def hipotenusa_getter(self):
        return self.__hipotenusa

    def cateto_a_setter(self, valor):
        self.__cateto_a = valor

    def cateto_b_setter(self, valor):
        self.__cateto_b = valor
````

````python
# Célula exemplo:
tri = TrianguloRetangulo(3, 4)

print(tri)
tri.cateto_a_setter(10)
print(tri)
````

Perceba que alteramos o valor do cateto A e ainda assim a hipotenusa manteve seu valor original. Isso é um erro comum ao lidar com objetos e entra na categoria dos erros de semântica, ou seja, você não será notificado pelo interpretador. Para resolver nosso problema, **Devemos atualizar o valor da hipotenusa sempre que alterarmos o valor de `cateto_a` e `cateto_b`.**

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """

    #adicionamos `__` antes dos nomes da variáveis para identificar que são privadas
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)

    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.__cateto_a, self.__cateto_b, self.__hipotenusa)

    def __atualiza_hipotenusa(self):
        self.__hipotenusa = (self.__cateto_a**2 + self.__cateto_b**2)**(1/2)

    def cateto_a_getter(self):
        return self.__cateto_a

    def cateto_b_getter(self):
        return self.__cateto_b

    def hipotenusa_getter(self):
        return self.__hipotenusa

    def cateto_a_setter(self, valor):
        self.__cateto_a = valor
        self.__atualiza_hipotenusa()

    def cateto_b_setter(self, valor):
        self.__cateto_b = valor
        self.__atualiza_hipotenusa()
````

````python
# Célula exemplo:
tri = TrianguloRetangulo(3, 4)
print(tri)
tri.cateto_a_setter(10)
print(tri)
````

Agora tudo parece funcionar corretamente. O segredo para a solução de problemas relacionados a classes é sempre ficar atento e lembrar que o depurador é seu melhor amigo. Isso se deve ao fato de que, em orientação a objetos, erros de semântica tendem a ser mais recorrentes, e mais graves, do que erros de sintaxe.
