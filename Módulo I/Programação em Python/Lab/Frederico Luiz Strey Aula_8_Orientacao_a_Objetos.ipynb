### 2.2. Herana

Ainda em nosso exemplo da classe `Carro`, podemos definir um automvel como herdeiro da classe `Veculos`. Mas como assim herdeira? Veja bem, se pensarmos nos vrios tipos de viculos (carros, caminhes, motos, vans, etc), todos eles possuem caractersticas em comum. Entre elas podemos citar:

Portanto, podemos definir uma classe mais genrica chamada `Veculo` da seguinte forma:

```
#Pseudocdigo
    classe Veculo:
        Propriedades:
            Ano,
            Placa,
            Proprietrio
        Mtodos:
            acelerar(),
            ligar_farol(),
            frear(),
            buzinar()
```

Podemos agora recriar a classe `Carro` como **herdeira** da classe `Veculo`:

```

    classe Carro (herdeira de Veculo):
        Mtodos:
            abrir_porta(),
            abrir_janela()
```

Perceba que no declaramos nenhuma propriedade. Com a herana, a classe `Carro` **herda** todos os mtodos e propriedades da classe `Veculo`. Tambm  possvel adicionar mtodos ou propriedades que no estejam definidas nas classe herdada, como  o caso de `carro.abrir_porta()` e `carro.abrir_janela()`.

Caso algum mtodo tenha o mesmo nome, mas funcione de maneira diferente (como por exemplo uma bicicleta no acende o farol da mesma forma que um carro ou at mesmo uma moto), basta apenas **sobrescrever** os mtodos que tero comportamento diferente, isto , declarar o mtodo novamente dentro da classe, com o mesmo nome da que precisa ser sobrescrita, e declarar as aes que servem aquele objeto.

*   ``__init__``  o nome reservado para o construtor de uma classe. Ele funciona parecido com uma funo e seus parmetros so respectivamente (*self*, *propriedade 1*, *propriedade 2*).

* A palavra "_self_" significa algo como "prprio" ou "auto". Ao trabalharmos com classes, usar esse termo como parmetro de uma funo indica que, quando aquela funo for chamada, propriedades do prprio objeto sero acessadas (podendo ser modificadas). Com a prtica vai tudo ficar bem claro!

* O `.`  usado para **acessar determinada caracterstica/propriedade** de algum objeto especificado.

Dessa forma, quando colocamos `self.nome`, estamos acessando o nome do prprio objeto onde a funo est declarada. Com isso, ao utilizar o init apresentado, sabemos que estamos modificando as propriedades `nome`, `ataque` e `vida` do nosso objeto, a fim de **inicializar** ele.

Vamos declarar nossa classe:

````python
#Execute essa clula para declarar a classe Personagem!
class Personagem:
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
    """
    def __init__(self, nome, ataque, vida):
        self.nome = nome
        self.ataque = ataque
        self.vida = vida
````

Agora, ficou muito mais simples exibir informaes sobre nossos personagens. Contudo, existe uma maneira melhor de implementar essa funcionalidade.

Estamos falando de um mtodo chamado `__str__`. Ao exibir um objeto com `print`, o _Python_ invoca esse mtodo, que acessa o objeto e as suas propriedades, retornando uma string no formato definido em `__str__`.

Alm de `__init__` e `__str__`, outros mtodos com nomes "especiais" podem melhorar sua classe; esses mtodos podem ser encontrados [aqui](https://docs.python.org/pt-br/3/library/functions.html).

````python
class Personagem:
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
    """
    def __init__(self, nome, ataque, vida):
        self.nome = nome
        self.ataque = ataque
        self.vida = vida

    def __str__(self):
        string_apresentacao = f"Ol! Meu nome  {self.nome}. Meu poder de ataque  {self.ataque} e tenho {self.vida} de vida!"
        return string_apresentacao
````

### 3.3. Herana

Agora que j vimos como funciona a criao de uma classe e seus mtodos, vejamos como isso pode ser aplicado com o conceito de herana.

Para isso, vamos imaginar uma classe mais especfica de personagem: a classe `Defensor`. Antes disso, vamos declarar nossa classe-me:

````python
class Personagem:
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
    """
    def __init__(self, nome, ataque, vida):
        self.nome = nome
        self.ataque = ataque
        self.vida = vida

    def __str__(self):
        string_apresentacao = f"Ol! Meu nome  {self.nome}. Meu poder de ataque  {self.ataque} e tenho {self.vida} de vida!"
        return string_apresentacao

    def esta_vivo(self):
        return self.vida > 0
````

Agora, podemos criar classes mais especficas, que herdem os mtodos e os atributos da classe `Personagem`. Alm dos atributos de um personagem genrico, a classe `Defensor` ter um escudo, o que lhe confere vida extra!

````python
class Defensor(Personagem):
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
        escudo: quantidade extra de vida(int ou float)
    """
    def __init__(self, nome, ataque, vida, escudo):
        super().__init__(nome, ataque, vida) #esta chamada serve para invocar o mtodo init da classe-me!
        self.escudo = escudo

    def __str__(self):
        return super().__str__() + f"Sou um defensor com {self.escudo} de escudo!" #Aqui, unimos  apresentao da classe-me informaes sobre o escudo
````

````python
gragas = Defensor("Gragas", 300, 5000, 2000)

if gragas.esta_vivo(): #repare que podemos usar mtodos da classe me sem declar-los dentro de "Defensor"!
    print(gragas)
else: #se a primeira linha for "gragas = Defensor("Gragas", 300, 0, 2000)", essa condicao sera satisfeita
    print(f"R.I.P {gragas.nome} :(")
````

No precisamos reescrever todos os mtodos de `Personagem` para o tipo `Defensor`, afinal ele _herdou_ todas as propriedades e mtodos de sua classe-me. Contudo, vale ressaltar que ao usarmos algum mtodo da classe-me na definio dos mtodos da classe-fillha, devemos lanar mo do termo "super()." seguido do nome desse mtodo - foi o que fizemos no init e no str. Por fim, repare que no h conflito ao usarmos funes de nomes iguais em classes que possuem relao de herana, pois  permitido *sobrescrever* funes.

**OBS:** Note que a classe herdeira, ou seja, a classe derivada, tem acesso a todas as propriedades e mtodos da classe principal, mas **o contrrio no  vlido**. A classe-me (no nosso exemplo, `Personagem`, no tem acesso  propriedade `escudo` da classe `Defensor`).
Pense como se fosse uma herana mesmo: a pessoa que recebeu a herana tem acesso a todo o dinheiro da pessoa que morreu, mas a pessoa que morreu no tem acesso a nada daquele que recebeu :(

**Exerccio 8.2.**

Agora que aprendemos sobre criao de mtodos, nos ajude a criar o mtodo `dobrar_vida`, que deve multiplicar a vida do nosso defensor por 2!

````python
class Defensor(Personagem):
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
        escudo: quantidade extra de vida(int ou float)
    """
    def __init__(self, nome, ataque, vida, escudo):
        super().__init__(nome, ataque, vida)
        self.escudo = escudo

    def __str__(self):
        return super().__str__() + f"Sou um defensor com {self.escudo} de escudo!"

    def dobrar_vida(self):
        self.vida *= 2
````

````python
class Personagem:
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
    """
    def __init__(self, nome, ataque, vida):
        self.nome = nome
        self.ataque = ataque
        self.vida = vida

    def __str__(self):
        string_apresentacao = f"Ol! Meu nome  {self.nome}. Meu poder de ataque  {self.ataque} e tenho {self.vida} de vida!"
        return string_apresentacao

    def recebe_dano(self, ataque):
        self.vida -= ataque
        if(self.vida < 0):
          self.vida = 0
    def esta_vivo(self):
        """
        Verifica se o personagem est vivo.
        """
        return self.vida > 0
````

A equipe contm um Atacante e um Defensor, que podemos enxergar como objetos de classes derivadas da classe Personagem. J conhecemos a classe Defensor, mas precisanos que ela seja um pouco mais robusta:

````python
class Defensor(Personagem):
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
        escudo: quantidade extra de vida(int ou float)
    """
    def __init__(self, nome, ataque, vida, escudo):
        super().__init__(nome, ataque, vida)
        self.escudo = escudo

    def __str__(self):
        return super().__str__() + f"Sou um defensor com {self.escudo} de escudo!"

    '''
    Essa funcao processa um ataque recebido de outro personagem levando
    em conta o escudo do defensor; para isso, sobrescrevemos a funo
    recebe_dano
    '''
    def recebe_dano(self, ataque):
        if self.escudo > 0:
          if (self.escudo - ataque) >= 0: #verifica se o escudo vai aguentar todo o dano recebido
              self.escudo -= ataque
          else: #caso em que o escudo aguentou apenas uma parte do dano
              self.vida -= (ataque - self.escudo)
              self.escudo = 0

              if(self.vida < 0):
                self.vida = 0
        else:
            self.vida -= ataque

            if(self.vida < 0):
                self.vida = 0
````

Agora, implemente a classe `Atacante`. Alm dos atributos de um personagem genrico, ele tem uma taxa (que varia de 0 a 1) de crtico, que representa uma porcentagem (referente  sua prpria fora) que deve ser adicionada em seu ataque.

Siga a seguinte estrutura:

Dica: Observe o exemplo da classe `Defensor` para realizar essa tarefa!

````python
class Atacante(Personagem):
    """
    Propriedades:
        nome : nome do personagem (string)
        forca : poder de ataque (int ou float)
        vida : quantidade de dano suportado, o hp (int ou float)
        taxa_critico: porcentagem (0 a 1) de chance crtica (float)
    """
    def __init__(self, nome, ataque, vida, taxa_critico):
        super().__init__(nome, ataque, vida)
        self.taxa_critico = taxa_critico

    def __str__(self):
        return super().__str__() + f" Tenho {self.taxa_critico * 100}% de chance de crtico!"

    def calcula_ataque(self):
        """
        Calcula o ataque levando em considerao a taxa de crtico.
        """
        ataque_critico = self.ataque * (1 + self.taxa_critico)
        return ataque_critico

    def recebe_dano(self, dano):
        """
        Recebe dano e atualiza a vida do atacante.
        Se a vida cair abaixo de zero, ela  ajustada para zero.
        """
        self.vida -= dano
        if self.vida < 0:
            self.vida = 0
````

### 4.2. Classe Equipe

Nossa classe Equipe dever ter um nome e dois personagens: um Atacante e um Defensor. Essa escolha facilita as coisas, mas poderamos fazer uma equipe com vrios personagens, sem distino entre Atacante e Defensor: bastaria que crissemos uma coleo de personagens! (lembre-se que uma coleo pode ser uma lista, tupla ou dicionrio).


Desse modo, a organizao dessa classe deve ser mais ou menos assim:
```
classe Equipe:
    Propriedades:
        nome,
        atacante,
        defensor
    Mtodos:
        construtor (init)
        impresso (str)
        foi_derrotada
        realiza_ataque
```

Obs.: A funo `foi_derrotada` deve verificar se existe algum personagem da equipe vivo (e, se sim, retornar `False`); j a `realiza_ataque` deve usar os ataques dos membros da equipe contra um personagem.

**Exerccio 8.4.**

Com base nas informaes dadas at aqui, complete nossa classe `Equipe`:

````python
class Equipe:
    """
    Classe para representar uma equipe de personagens.

    Propriedades:
        nome: Nome da equipe (str)
        atacante: Personagem do tipo Atacante
        defensor: Personagem do tipo Defensor
    """

    def __init__(self, nome: str, atacante: Atacante, defensor: Defensor):
        self.nome = nome
        self.atacante = atacante
        self.defensor = defensor

    def __str__(self) -> str:
        return f"Equipe {self.nome}: {self.atacante.nome} (Atacante) e {self.defensor.nome} (Defensor)"

    def foi_derrotada(self) -> bool:
        """
        Verifica se a equipe foi derrotada, ou seja, se no h personagens vivos.
        """
        return self.atacante.vida <= 0 and self.defensor.vida <= 0

    def realiza_ataque(self, alvo: Personagem) -> None:
        """
        Realiza um ataque do atacante da equipe contra um alvo.
        """
        if self.atacante.esta_vivo():
            ataque = self.atacante.calcula_ataque()
            alvo.recebe_dano(ataque)
            print(f"{self.atacante.nome} atacou {alvo.nome} e causou {ataque:.2f} de dano!")
        else:
            print(f"{self.atacante.nome} no pode atacar porque est derrotado!")
````

**Exerccio 8.5.**

Implemente o mtodo `realiza_ataques`. Para isso, leve em conta que os ataques s devem ser direcionados ao atacante da outra equipe se o defensor estiver morto.

````python
class Equipe:
    """
    Classe para representar uma equipe de personagens.

    Propriedades:
        nome: Nome da equipe (str)
        atacante: Personagem do tipo Atacante
        defensor: Personagem do tipo Defensor
    """

    def __init__(self, nome: str, atacante: Atacante, defensor: Defensor):
        self.nome = nome
        self.atacante = atacante
        self.defensor = defensor

    def __str__(self) -> str:
        return f"{self.nome} - Atacante: {self.atacante.nome} (Vida: {self.atacante.vida}) | Defensor: {self.defensor.nome} (Vida: {self.defensor.vida}, Escudo: {self.defensor.escudo})"

    def foi_derrotada(self) -> bool:
        """
        Verifica se a equipe foi derrotada, ou seja, se no h personagens vivos.
        """
        return self.atacante.vida <= 0 and self.defensor.vida <= 0

    def realiza_ataque(self, alvo: Personagem) -> None:
        """
        Realiza um ataque do atacante da equipe contra um alvo.
        """
        if self.atacante.esta_vivo():
            ataque = self.atacante.calcula_ataque()
            alvo.recebe_dano(ataque)
            print(f"{self.atacante.nome} atacou {alvo.nome} e causou {ataque:.2f} de dano!")
        else:
            print(f"{self.atacante.nome} no pode atacar porque est derrotado!")

    def realiza_ataques(self, outra_equipe: 'Equipe') -> None:
        """
        Realiza ataques contra a outra equipe.
        Se o defensor da outra equipe estiver vivo, o ataque  direcionado a ele.
        Se o defensor estiver derrotado, o ataque  direcionado ao atacante da outra equipe.
        """
        if outra_equipe.defensor.esta_vivo():
            print(f"{self.nome} ataca o defensor {outra_equipe.defensor.nome} da equipe {outra_equipe.nome}!")
            self.realiza_ataque(outra_equipe.defensor)
        else:
            print(f"{self.nome} ataca o atacante {outra_equipe.atacante.nome} da equipe {outra_equipe.nome}!")
            self.realiza_ataque(outra_equipe.atacante)

````

 essa altura, j criamos um projeto bem legal. Vamos testar:

````python
if __name__ == "__main__":
    # Criao de personagens
    atacante1 = Atacante("Legolas", ataque=15, vida=100, taxa_critico=0.5)
    defensor1 = Defensor("Gimli", ataque=10, vida=120, escudo=50)

    atacante2 = Atacante("Nazgl", ataque=15, vida=100, taxa_critico=0.5)
    defensor2 = Defensor("Orc", ataque=5, vida=70, escudo=20)

    # Criao das equipes
    equipe1 = Equipe("Sociedade do Anel", atacante1, defensor1)
    equipe2 = Equipe("Exrcito das Sombras", atacante2, defensor2)

    # Impresso das equipes
    print(equipe1)
    print(equipe2)

    # Realiza ataques entre as equipes
    while not equipe1.foi_derrotada() and not equipe2.foi_derrotada():
        equipe1.realiza_ataques(equipe2)
        if not equipe2.foi_derrotada():
            equipe2.realiza_ataques(equipe1)

    # Verificar qual equipe foi derrotada
    if equipe1.foi_derrotada():
        print(f"A equipe {equipe1.nome} foi derrotada!")
    if equipe2.foi_derrotada():
        print(f"A equipe {equipe2.nome} foi derrotada!")
````

## 5. O que pode dar errado?

Como pudemos perceber, nossos cdigos ficaram um pouco mais complexos - com isso, surgem novos possveis erros. O exemplo abaixo visa ilustrar isso:

A classe abaixo representa um tringulo retngulo. Como bem sabemos, um tringulo retngulo de catetos com, respectivamente, $a$ e $b$ de comprimento e hipotenusa de $c$ de comprimento, deve seguir a frmula:
$c^2 = a^2 + b^2$

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """

    #adicionamos `__` antes dos nomes da variveis para identificar que so privadas
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)

    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.__cateto_a, self.__cateto_b, self.__hipotenusa)

    def cateto_a_getter(self):
        return self.__cateto_a

    def cateto_b_getter(self):
        return self.__cateto_b

    def hipotenusa_getter(self):
        return self.__hipotenusa

    def cateto_a_setter(self, valor):
        self.__cateto_a = valor

    def cateto_b_setter(self, valor):
        self.__cateto_b = valor
````

````python
# Clula exemplo:
tri = TrianguloRetangulo(3, 4)

print(tri)
tri.cateto_a_setter(10)
print(tri)
````

Perceba que alteramos o valor do cateto A e ainda assim a hipotenusa manteve seu valor original. Isso  um erro comum ao lidar com objetos e entra na categoria dos erros de semntica, ou seja, voc no ser notificado pelo interpretador. Para resolver nosso problema, **Devemos atualizar o valor da hipotenusa sempre que alterarmos o valor de `cateto_a` e `cateto_b`.**

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """

    #adicionamos `__` antes dos nomes da variveis para identificar que so privadas
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)

    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.__cateto_a, self.__cateto_b, self.__hipotenusa)

    def __atualiza_hipotenusa(self):
        self.__hipotenusa = (self.__cateto_a**2 + self.__cateto_b**2)**(1/2)

    def cateto_a_getter(self):
        return self.__cateto_a

    def cateto_b_getter(self):
        return self.__cateto_b

    def hipotenusa_getter(self):
        return self.__hipotenusa

    def cateto_a_setter(self, valor):
        self.__cateto_a = valor
        self.__atualiza_hipotenusa()

    def cateto_b_setter(self, valor):
        self.__cateto_b = valor
        self.__atualiza_hipotenusa()
````

````python
# Clula exemplo:
tri = TrianguloRetangulo(3, 4)
print(tri)
tri.cateto_a_setter(10)
print(tri)
````
