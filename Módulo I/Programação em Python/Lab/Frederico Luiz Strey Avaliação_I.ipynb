## Exercício 6.1.

Dados os pontos cartesianos $(X_1, Y_1)$ e $(X_2, Y_2)$ como entrada pelo teclado, escreva um código que exiba na tela a distância entre ambos.

**Observação:** A distância $AB$ entre dois pontos $A = (X_a, Y_a)$ e $B = (X_b, Y_b)$ é dada por $|AB| = \sqrt{{(X_b - X_a)}^{2} + {(Y_b - Y_a)}^{2}}$.

**Entrada:**\
0\
0\
1\
1\
**Saída:**\
1.4142135623730951

````python
import math
def calcular_distancia():
  x1 = float(input('digite x1'))
  y1 = float(input('digite y1'))
  x2 = float(input('digite x2'))
  y2 = float(input('digite y2'))
  distancia = math.sqrt((x2-x1)**2+(y2-y1)**2)
  print(distancia)
calcular_distancia()
````

## Exercício 6.2.

Escreva um código que receba duas string, a primeira sendo uma série de caracteres e a segunda a palavra que você deseja encontrar naquela sequência (o código deve desconsiderar diferenças entre letras maiúsculas e minúsculas e deve imprimir a palavra encontrada em CAIXA ALTA):

**Entrada:**\
abksjdIntrocompaksdj\
introcomp\
**Saída:**\
Palavra encontrada: INTROCOMP\
\
**Entrada:**\
abksabksabks\
introcomp\
**Saída:**\
Palavra não encontrada!

````python
def decifrar_mensagem(string,palavra):
  string_minuscula = string.lower()
  palavra_minuscula = palavra.lower()
  if palavra_minuscula in string_minuscula:
        print(f'Palavra encontrada: {palavra.upper()}')
  else:
        print('Palavra não encontrada!')
entrada1 = "abksjdIntrocompaksdj"
palavra1 = "introcomp"
decifrar_mensagem(entrada1, palavra1)

entrada2 = "abksabksabks"
palavra2 = "introcomp"
decifrar_mensagem(entrada2, palavra2)
````

## Exercício 6.3.
Escreva um código que receba como entrada dois números $(N, M)$ e, na saída, exiba na tela `N eh multiplo de M` se $N$ for múltiplo de $M$ e, caso contrário, exiba `N nao eh multiplo de M`.

**Entrada:**\
20\
2\
**Saída:**\
20 eh multiplo de 2\
\
**Entrada:**\
21\
2\
**Saída:**\
21 nao eh multiplo de 2

````python
def verificar_multiplo(N,M):
    if N % M == 0:
        print(f"{N} eh multiplo de {M}")
    else:
        print(f"{N} nao eh multiplo de {M}")

entrada1 = verificar_multiplo(20,2)
entrada2 = verificar_multiplo(21,2)
````

## Exercício 6.4.
Vamos escrever um programa que calcula a quantia total de dinheiro em uma conta após uma sequência de operações. Teremos 3 operações:

- D -> Depositar
- S -> Sacar
- E -> Encerrar


| Entrada: | Saída |
| :------- | :----: |
|D<br>300<br>D<br>300<br>S<br>200<br>D<br>100<br>E<br>| 500 |



````python
def calcular_conta():
    saldo = 0.0

    while True:
        print(f"Saldo Atual : R${saldo:.2f}")
        operacao = input("Digite a operação (D - Depositar, S - Sacar, E - Encerrar): ").strip().upper()

        if operacao == 'D':
            valor= float(input("Digite o valor a ser depositado: "))
            saldo = saldo + valor
            print(f"Depósito de R${valor:.2f} realizado com sucesso. Saldo atual R${saldo:.2f}")


        elif operacao == 'S':
          valor = float(input("Digite o valor a ser sacado: "))
          if valor > saldo:
            print("Saldo insuficiente para saque.")
          else:
            saldo -= valor
            print(f"Saque de R${valor:.2f} realizado com sucesso. Saldo atual de R${saldo:.2f} ")


        elif operacao == 'E':
            print(f"Operações encerradas. Saldo final: R${saldo:.2f}")
            break

        else:
            print("Operação inválida! Por favor, escolha D, S ou E.")

# Chamada da função
calcular_conta()
````
