```python
# Garantir que a coluna 'OrderDateKey' esteja no formato de data
df['OrderDateKey'] = pd.to_datetime(df['OrderDateKey'], format='%Y%m%d')

# Criar a coluna 'Mês_Ano' com o formato 'YYYY-MM'
df['Mês_Ano'] = df['OrderDateKey'].dt.to_period('M')

# Agrupar os dados por 'Mês_Ano' e calcular a quantidade de clientes únicos (CustomerKey) por mês
clientes_unicos_por_mes = df.groupby('Mês_Ano')['CustomerKey'].nunique()

# Calcular a frequência média de clientes únicos por mês
frequencia_media_clientes_unicos = clientes_unicos_por_mes.mean()
```

```python

# Garantir que a coluna 'OrderDateKey' esteja no formato de data
dff['OrderDateKey'] = pd.to_datetime(dff['OrderDateKey'], format='%Y%m%d')

# Agrupar os dados por 'OrderDateKey' e 'CustomerKey' e contar a frequência de cada cliente
frequencia_clientes_por_data = dff.groupby(['OrderDateKey', 'CustomerKey']).size().reset_index(name='Frequencia')

# Verificar as primeiras linhas do DataFrame
frequencia_clientes_por_data
```

```python
frequencia_clientes_por_data['Mês-Ano'] = frequencia_clientes_por_data['OrderDateKey'].dt.strftime('%m-%Y')
frequencia_clientes_por_data
```

```python
media_frequencia_por_data = frequencia_clientes_por_data.groupby('Mês-Ano')['Frequencia'].mean().reset_index(name='Média_Frequencia')
media_frequencia_por_data
```

```python
media_frequencia_por_data['Mês-Ano'] = pd.to_datetime(media_frequencia_por_data['Mês-Ano'], format='%m-%Y')
media_frequencia_por_data = media_frequencia_por_data.sort_values(by='Mês-Ano').reset_index(drop=True)
media_frequencia_por_data
```
